<h1>7) Curvilinear Coordinates</h1>

<h4>Notes / Bibliography : Taken from Shaums Vector Analaysis, Spiegel. to be continued with expansion from other sources</h4> 

<h2>Transformation of coordinates:</h2> 

Let the rectangular coordinates $(x,\ y,\ z) $ of any point be expressed as functions of $(u_1, u_2, u_3) $ so that <br><br>


1) $ \qquad    x = x(u_1,\ u_2,\ u_3),\qquad     y = y(u_1,\ u_2,\ u_3),\qquad   z= z(u_1,\ u_2,\ u_3) $


<br><br>
suppose that 1) can be solved for $ (u_1, u_2, u_3) $ in terms of $ (x, y, z) $ <br><br>



2) $ \qquad    u_1 = u_1(x,\ y,\ z),\qquad      u_2 = u_2(x,\ y,\ z),\qquad     u_3 = u_3(x,\ y,\ z)$


<p>the functions in <b>1)</b> and <b>2)</b> are assumed to be single-valued and to have continuous derivatives so that the correspondance between $(x,\ y,\ z)$ and $(u_1,\ u_2,\ u_3)$is unique.</p>
<br>
<p>In practice this assumption may not apply at certain points and special consideration is required.</p>
<br>
<p>Given point P with rectangular coordinates $ (x,\ y,\ z) $ we can from <b>2)</b> associate a unique set of coordinates $(u_1,\ u_2,\ u_3) $ called the curvilinear coordinates of P. The sets of equations 1) and 2) define a transformation of coordinates.</p>

<h2>Orthogonal Curvilinear Coordinates</h2> 

<p>The surfaces $ u_1 = c_1$, $ u_2 = c_2 $ and $ u_3 = c_3 $ where$ c_1, c_2, c_3 $ are constants are called coordinate surfaces and each pair of these surfaces intersect in curves called coordinate curves or lines. If the coordinate system is called orthogonal. The $ (u_1,\ u_2,\ u_3) $  coordinate cutves of a curvilinear system are analogous to the $ (x,\ y,\ z) $ axes of a rectangular system.</p>



<h2>Unit Vectors in Curvilinear Systems</h2> 

Let $ \mathbf{r} = x \mathbf{i} + y \mathbf{j} + z \mathbf{k} $ be the position vector of a point P. Then <b>1)</b> can be written $ \mathbf{r}=r(u_1,u_2,u_3) $, A tangent vector to the $ u_1 $ curve at P (for which $ u_2 $ and $u_3 $ are constants) is $ \frac{dr}{du_1} $. Then a unit tangent vector in this direction is 


$$
\displaystyle\mathbf e_1 = \frac{\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_1} }{\left |\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_1} \right |} \ \ so\ that\ \frac{\mathrm{d}\mathbf r }{\mathrm{d} u_1}=h_1 \mathbf e_1 \ where \ \ h_1 = \left |\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_1} \right |  
$$


Similarly if $ \mathbf e_2 $ and $ \mathbf e_2 $ are unit tangent vectors to the $ u_2 $ and $ u_3 $ curves at P respectively then 


$$
\displaystyle\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_2}=h_2\ \mathbf e_2 \quad \ where \quad \ h_2 = \left |\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_2} \right |\qquad\qquad\displaystyle\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_3}=h_3\ \mathbf e_3 \quad \ where \quad \ h_3 = \left |\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_3} \right |
$$


The quantities $h_1,\ h_2,\ h_3$ are called scale factors. The unit vectors $ e_1,\ e_2,\ e_3 $ are in the  directions of increasing $ u_1,\ u_2,\ u_3 $ respectively.

Since $ \mathbf{grad} u_1 $ is a vector at P normal to the surface $ u_1 = c_1 $, a unit vector in this direction is given by



$$\mathbf{ E_1} = \frac{\mathbf{grad}\ u_1} {  | \mathbf{ grad }\ u_1  | }=\frac{\nabla u_1}{| \nabla u_1  |}$$



 Similarly 


 $$ \mathbf E_2 =\frac{\nabla u_2}{\left| \nabla u_2 \right |}  \\$$



 $$ \mathbf E_3 =\frac{\nabla u_3}{\left| \nabla u_3 \right |} \\$$



at P are normal to the surfaces $ u_2=c_2 $ and $ u_3=c_3 $ respectively.

Thus at each point P of a curvilinear coordinate system there exist in general two sets of unit vectors, $ \mathbf{e_1,\ e_2,\ e_3} $ tangent to the coordinate curves and $ \mathbf{E_1, E_2, E_3} $ normal to the coordinate surfaces. The sets become identical if and only if the curvilinear coordinate system is orthogonal. 

Both sets are analogous to the $ \mathbf{i,\ j,\ k} $ vectors in rectangular coordinates but are unlike them in that they may change directions from point to point. 

It can be shown that the sets 



$$\displaystyle\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_1},\ \frac{\mathrm{d}\mathbf r }{\mathrm{d} u_2},\ \frac{\mathrm{d}\mathbf r }{\mathrm{d} u_3}$$      and


$$
\nabla u_1, \ \nabla u_2,\ \nabla u_3
$$


constitute reciprocal systems of vectors. 
    

A vector can be represented in terms of the unit base vectors $ \mathbf{e_1,\ e_2,\ e_3} $ or $\mathbf{E_1,E_2,E_3} $ in this form


$$
A = A_1 \mathbf{e_1} + A_2 \mathbf{e_2} + A_3 \mathbf{e_3} = a_1 \mathbf{E_1} + a_2 \mathbf{E_2} + a_3 \mathbf{E_3} 
$$


where the $ A_i $ and $ a_i $are the respective components of <b>A</b> in each system.


We can also represent <b>A</b> in terms of the base vectors  


$$
\displaystyle\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_1},\ \frac{\mathrm{d}\mathbf r }{\mathrm{d} u_2},\ \frac{\mathrm{d}\mathbf r }{\mathrm{d} u_3}        \quad or \quad         \nabla u_1, \ \nabla u_2,\ \nabla u_3
$$


which are called unitary base vectors but are not unit vectors.




$$
\displaystyle\mathbf A = C_1\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_1} + C_2 \frac{\mathrm{d}\mathbf r }{\mathrm{d} u_2} + C_3 \frac{\mathrm{d}\mathbf r }{\mathrm{d} u_3} = C_1 \alpha_1 + C_2 \alpha_2 + C_3 \alpha_3
$$

and     


$$
\mathbf A =c_1 \nabla u_1 + c_2 \nabla u_2 + c_3 \nabla u_3 = c_1 \beta_1 + c_2 \beta_2 + c_3 \beta_3
$$

where $ C_1,\ C_2,\ C_3 $are called the <b>contravariant components</b> of <b>A</b> and $ c_1,\ c_2,\ c_3 $ are the <b>covariant components</b> of <b>A</b>. 


$$
\alpha_\rho = \frac{\partial \mathbf r}{\partial u_\rho}   ,\qquad          \beta_\rho = \nabla u_\rho,\qquad          \rho =1,\ 2,\ 3         
 $$ 


<h2>Arc Length and Volume Elements</h2>



From \( \mathbf r = \mathbf r(u_1,\ u_2,\ u_3) \) we have


$$\displaystyle d\mathbf r = \frac{\mathrm{d}\mathbf r }{\mathrm{d} u_1}d u_1 +\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_2}d u_2 + \frac{\mathrm{d}\mathbf r }{\mathrm{d} u_3} d u_3=h_1 d u_1 \mathbf e_1+h_2 d u_2 \mathbf e_2+h_3 d u_3 \mathbf e_3$$

Then the differential arc length ds is determined from $ds_2 = d \mathbf{r}\cdot d \mathbf{r} $. For orthogonal systems,


$$\mathbf e_1.\mathbf e_2=\mathbf e_2.\mathbf e_3=\mathbf e_3.\mathbf e_1=0\quad   and\\ds^2=h_1^2d u_1^2+h_2^2d u_2^2+h_3^2d u_3^2$$

Along a $ u_1 $ curve, $ u_2 $ and $ u_3 $ are constants so that $ d \mathbf{r} = h_1 du_1 \mathbf{e_1} $. Then the differential arc length $ds_1 $ along $u_1 $ at P is $ h_1 du_1 $. Similarly the differential arc length along $ u_2 $ and $ u_3 $ at P are $ds_2 = h_2 du_2 $ and $ds_3 = h_3 du_3 $ <br>


    The volume element for an orthogonal curvilinear system is 





$$ d V=|(h_1 d u_1 \mathbf e_1).(h_2 d u_2 \mathbf e_2) \times h_3 d u_3 \mathbf e_3|=h_1 h_2 h_3\ d u_1 d u_2 d u_3\quad $$ since $$\quad |\mathbf e_1.\mathbf e_2 \times \mathbf e_3| = 1 $$

<h2>The Gradient, Divergence and Curl</h2>

can be expressed in terms of curvilinear coordinates. If $ \phi $ is a scalar function and 

$ \displaystyle \mathbf{A} = A_1 \mathbf{e_1} + A_2 \mathbf{e_2} + A_3 \mathbf{e_3} $ a vector function of orthogonal curvilinear coordinates $ u_1,\ u_2,\ u_3$ then the following is valid
<br>
<br>
1) $ \displaystyle \nabla \Phi = \mathbf{grad}\ \Phi = \frac{1}{h_1}\frac{d\Phi}{du_1}\mathbf e_1+ \frac{1}{h_2}\frac{d\Phi}{du_2}\mathbf e_2+ \frac{1}{h_3}\frac{d\Phi}{du_3}\mathbf e_3
$
<br>
<br>
2) $\displaystyle \nabla\cdot\mathbf A=\mathbf{div}\ \mathbf A=\frac{1}{h_1 h_2 h_3}\left [ \frac{\partial }{\partial u_1}(h_2 h_3 A_1)+\frac{\partial }{\partial u_2}(h_3 h_1 A_2)+\frac{\partial }{\partial u_3}(h_1 h_2 A_3) \right ]
$
<br>
<br>
3) $ \displaystyle \nabla \times \mathbf A = \mathbf{curl}\ \mathbf A = \frac{1}{h_1 h_2 h_3}\begin{vmatrix} h_1\mathbf e_1 & h_2\mathbf e_2 &h_3\mathbf e_3 \\ \frac{\partial }{\partial u_1} & \frac{\partial }{\partial u_2} & \frac{\partial }{\partial u_3} \\ h_1 A_1 &h_2 A_2 &h_3 A_3 \end{vmatrix} 
$
<br>
  <br>  
4) $ \displaystyle \nabla^2 \Phi = Laplacian\ of\ \Phi = \frac{1}{h_1 h_2 h_3}\left[ \frac{\partial }{\partial u_1}\left ( \frac{h_2 h_3}{h_1} \frac{\partial \Phi}{\partial u_1}\right ) + \frac{\partial }{\partial u_2}\left ( \frac{h_3 h_1}{h_2} \frac{\partial \Phi}{\partial u_2}\right ) + \frac{\partial }{\partial u_3}\left ( \frac{h_1 h_2}{h_3} \frac{\partial \Phi}{\partial u_3}\right ) \right]
$
<br>
<br>
if $ h_1 = h_2 = h_3 = 1$ and $\mathbf e_1 \mathbf e_2 \mathbf e_3 $ are replaced by $\mathbf i \mathbf j \mathbf k$ these reduce to the usual expressions in rectangular coordinates where $ (u_1,\ u_2,\ u_3)$ are replaced by $(x, y, z)$ .

<h1 style="color:red">Example: Contravariant Vector</h1>
<h4>note: these examples lead to the subject of tensor analysis, useful in General Relativity </h4>
    If $\mathbf A$is a given vector defined wiithin two general curvilinear coordinate systems $(u_1,\ u_2,\ u_3)$ and $(\bar u_1,\ \bar u_2,\ \bar u_3)$ then to find the relation between the contravariant
    components the calculation is as follows :-<br>
    <br>
    $$
    x_1= x_1(u_1,\ u_2,\ u_3), \qquad x_2=x_2(u_1,\ u_2,\ u_3)\qquad x_3=x_3(u_1,\ u_2,\ u_3)
    $$
    and 
    $$
    x_1= x_1(\bar u_1,\ \bar u_2,\ \bar u_3), \qquad x_2=x_2(\bar u_1,\ \bar u_2,\ \bar u_3)\qquad x_3=x_3(\bar u_1,\ \bar u_2,\ \bar u_3)
    $$
    <br>
    Then there exists a transformation between the $(u_1,\ u_2,\ u_3)$ system and the $(\bar u_1,\ \bar u_2,\ \bar u_3)$ system defined by <br>
    $$
    u_1= u_1(\bar u_1,\ \bar u_2,\ \bar u_3), \qquad u_2=u_2(\bar u_1,\ \bar u_2,\ \bar u_3)\qquad u_3=u_3(\bar u_1,\ \bar u_2,\ \bar u_3)\qquad\qquad (2)
    $$
    <br> and conversely. Now <br>
    $$
    d\mathbf r = \displaystyle \frac{\partial \mathbf r}{\partial u_1}du_1 + \frac{\partial \mathbf r}{\partial u_2}du_2 + \frac{\partial \mathbf r}{\partial u_3}du_3 = \mathbf{\alpha}_1 du_1+\mathbf{\alpha}_2 du_2+\mathbf{\alpha}_3 du_3
    $$
    <br> and <br>
    $$
    d\mathbf r = \displaystyle \frac{\partial \mathbf r}{\partial \bar u_1}d\bar u_1 + \frac{\partial \mathbf r}{\partial \bar u_2}d\bar u_2 + \frac{\partial \mathbf r}{\partial \bar u_3}d\bar u_3= \mathbf{\bar\alpha}_1 d\bar u_1+\mathbf{\bar\alpha}_2 d\bar u_2+\mathbf{\bar\alpha}_3 d\bar u_3
    $$    
    <br> then <br>
    $$
    \mathbf{\alpha}_1 du_1+\mathbf{\alpha}_2 du_2+\mathbf{\alpha}_3 du_3 = \mathbf{\bar\alpha}_1 d\bar u_1+\mathbf{\bar\alpha}_2 d\bar u_2+\mathbf{\bar\alpha}_3 d\bar u_3 \qquad\qquad (3)
    $$
    <br>
    and from (2) the differentials are <br>
    $$
    d u_1 = \displaystyle \frac{\partial u_1}{\partial \bar u_1}d\bar u_1 + \frac{\partial u_1}{\partial \bar u_2}d\bar u_2+\frac{\partial u_1}{\partial \bar u_3}d\bar u_3$$
   $$ d u_2 = \displaystyle \frac{\partial u_2}{\partial \bar u_1}d\bar u_1 + \frac{\partial u_2}{\partial \bar u_2}d\bar u_2+\frac{\partial u_2}{\partial \bar u_3}d\bar u_3$$
   $$ d u_3 = \displaystyle \frac{\partial u_3}{\partial \bar u_1}d\bar u_1 + \frac{\partial u_3}{\partial \bar u_2}d\bar u_2+\frac{\partial u_3}{\partial \bar u_3}d\bar u_3 $$
    
These can be substituted into (3) <br>
    $$\mathbf{\alpha}_1 \left(\displaystyle \frac{\partial u_1}{\partial \bar u_1}d\bar u_1 + \frac{\partial u_1}{\partial \bar u_2}d\bar u_2+\frac{\partial u_1}{\partial \bar u_3}d\bar u_3\right)+\mathbf{\alpha}_2 \left(  \frac{\partial u_2}{\partial \bar u_1}d\bar u_1 + \frac{\partial u_2}{\partial \bar u_2}d\bar u_2+\frac{\partial u_2}{\partial \bar u_3}d\bar u_3 \right)+\mathbf{\alpha}_3 \left(\frac{\partial u_3}{\partial \bar u_1}d\bar u_1 + \frac{\partial u_3}{\partial \bar u_2}d\bar u_2+\frac{\partial u_3}{\partial \bar u_3}d\bar u_3\right) = \mathbf{\bar\alpha}_1 d\bar u_1+\mathbf{\bar\alpha}_2 d\bar u_2+\mathbf{\bar\alpha}_3 d\bar u_3 \qquad\qquad (3)$$
    
and rearranged <br>
    $$\mathbf{\alpha}_1 \frac{\partial u_1}{\partial \bar u_1}d\bar u_1 + \mathbf{\alpha}_1 \frac{\partial u_1}{\partial \bar u_2}d\bar u_2+\mathbf{\alpha}_1\frac{\partial u_1}{\partial \bar u_3}d\bar u_3+ \mathbf{\alpha}_\frac{\partial u_2}{\partial \bar u_1}d\bar u_1 + \mathbf{\alpha}_2\frac{\partial u_2}{\partial \bar u_2}d\bar u_2+\mathbf{\alpha}_2\frac{\partial u_2}{\partial \bar u_3}d\bar u_3 + \mathbf{\alpha}_3\frac{\partial u_3}{\partial \bar u_1}d\bar u_1 + \mathbf{\alpha}_3\frac{\partial u_3}{\partial \bar u_2}d\bar u_2+\mathbf{\alpha}_3\frac{\partial u_3}{\partial \bar u_3}d\bar u_3 = \mathbf{\bar\alpha}_1 d\bar u_1+\mathbf{\bar\alpha}_2 d\bar u_2+\mathbf{\bar\alpha}_3 d\bar u_3$$
so <br><br>   
$$\displaystyle \left( \mathbf{\alpha}_1 \frac{\partial u_1}{\partial \bar u_1}+ \mathbf{\alpha}_2 \frac{\partial u_2}{\partial \bar u_1}+\mathbf{\alpha}_3\frac{\partial u_3}{\partial \bar u_1}\right)d\bar u_1+\left(\mathbf{\alpha}_1 \frac{\partial u_1}{\partial \bar u_2}+\mathbf{\alpha}_2\frac{\partial u_2}{\partial \bar u_2}+\mathbf{\alpha}_3\frac{\partial u_3}{\partial \bar u_2}\right)d\bar u_2+\left(\mathbf{\alpha}_1\frac{\partial u_1}{\partial \bar u_3}+\mathbf{\alpha}_2\frac{\partial u_2}{\partial \bar u_3} + \mathbf{\alpha}_3\frac{\partial u_3}{\partial \bar u_3}\right)d\bar u_3 = \mathbf{\bar\alpha}_1 d\bar u_1+\mathbf{\bar\alpha}_2 d\bar u_2+\mathbf{\bar\alpha}_3 d\bar u_3   $$
    
Therefore if the equality is true, <br>
    
$$\mathbf{\bar\alpha}_1 =  \displaystyle \mathbf{\alpha}_1 \frac{\partial u_1}{\partial \bar u_1}+ \mathbf{\alpha}_2 \frac{\partial u_2}{\partial \bar u_1}+\mathbf{\alpha}_3\frac{\partial u_3}{\partial \bar u_1}$$
    
$$\mathbf{\bar\alpha}_2 =  \displaystyle \mathbf{\alpha}_1 \frac{\partial u_1}{\partial \bar u_2}+ \mathbf{\alpha}_2 \frac{\partial u_2}{\partial \bar u_2}+\mathbf{\alpha}_3\frac{\partial u_3}{\partial \bar u_2}$$

$$\mathbf{\bar\alpha}_3 =  \displaystyle \mathbf{\alpha}_1 \frac{\partial u_1}{\partial \bar u_3}+ \mathbf{\alpha}_2 \frac{\partial u_2}{\partial \bar u_3}+\mathbf{\alpha}_3\frac{\partial u_3}{\partial \bar u_3}$$
   
These relations are equations (4). 
    <br> remembering that $\alpha_1 = \frac{\partial \mathbf r}{\partial u_1} = h_1 \mathbf e_1$ where $\mathbf e_1$ is a unit vector in the curvilinear system $(u_1,\ u_2,\ u_3)$ in the direction of increasing $u_1$. Similarly for the other components, we find that the vector can be expressed in either system as <br>
$$\mathbf A = C_1 \mathbf \alpha_1 + C_2 \mathbf \alpha_2 + C_3 \mathbf \alpha_3 = \bar C_1 \mathbf{\bar \alpha_1} +  \bar C_2 \mathbf{\bar \alpha_2} +  \bar C_3 \mathbf{\bar \alpha_3} \qquad\qquad(5)$$
    
where $C_1,\ C_2,\ C_3$ and $\bar C_1,\ \bar C_2,\ \bar C_3$ are the contravariant components of $\mathbf A$ in the two systems. Substituting the equations from (4) into (5) <br>
$$ \bar C_1 \mathbf{\bar \alpha_1} +  \bar C_2 \mathbf{\bar \alpha_2} +  \bar C_3 \mathbf{\bar \alpha_3} =\bar C_1 \left(\mathbf{\alpha}_1 \frac{\partial u_1}{\partial \bar u_1}+ \mathbf{\alpha}_2 \frac{\partial u_2}{\partial \bar u_1}+\mathbf{\alpha}_3\frac{\partial u_3}{\partial \bar u_1}\right)+  \bar C_2 \left(\mathbf{\alpha}_1 \frac{\partial u_1}{\partial \bar u_2}+ \mathbf{\alpha}_2 \frac{\partial u_2}{\partial \bar u_2}+\mathbf{\alpha}_3\frac{\partial u_3}{\partial \bar u_2}\right) +  \bar C_3 \left( \mathbf{\alpha}_1 \frac{\partial u_1}{\partial \bar u_3}+ \mathbf{\alpha}_2 \frac{\partial u_2}{\partial \bar u_3}+\mathbf{\alpha}_3\frac{\partial u_3}{\partial \bar u_3}\right)$$
and grouping terms again <br>
$$=\bar C_1 \mathbf{\alpha}_1 \frac{\partial u_1}{\partial \bar u_1}+ \bar C_1\mathbf{\alpha}_2 \frac{\partial u_2}{\partial \bar u_1}+\bar C_1\mathbf{\alpha}_3\frac{\partial u_3}{\partial \bar u_1}+   \bar C_2\mathbf{\alpha}_1 \frac{\partial u_1}{\partial \bar u_2}+ \bar C_2\mathbf{\alpha}_2 \frac{\partial u_2}{\partial \bar u_2}+\bar C_2\mathbf{\alpha}_3\frac{\partial u_3}{\partial \bar u_2} +   \bar C_3 \mathbf{\alpha}_1 \frac{\partial u_1}{\partial \bar u_3}+ \bar C_3\mathbf{\alpha}_2 \frac{\partial u_2}{\partial \bar u_3}+\bar C_3\mathbf{\alpha}_3\frac{\partial u_3}{\partial \bar u_3}$$
    
$$=\left(\bar C_1  \frac{\partial u_1}{\partial \bar u_1}+\bar C_2\frac{\partial u_1}{\partial \bar u_2}+  \bar C_3  \frac{\partial u_1}{\partial \bar u_3}\right) \mathbf{\alpha}_1+ \left(\bar C_1 \frac{\partial u_2}{\partial \bar u_1}+ \bar C_2 \frac{\partial u_2}{\partial \bar u_2}+\bar C_3 \frac{\partial u_2}{\partial \bar u_3}\right)\mathbf{\alpha}_2+\left(\bar C_1\frac{\partial u_3}{\partial \bar u_1}+  \bar C_2\frac{\partial u_3}{\partial \bar u_2} +\bar C_3\frac{\partial u_3}{\partial \bar u_3} \right)\mathbf{\alpha}_3$$
    
<br> after the tedious and time consuming process, we find the relations between the contravariant components by insepection, as before <br>
$$ C_1 = \displaystyle \bar C_1 \frac{\partial u_1}{\partial \bar u_1}+\bar C_2 \frac{\partial u_1}{\partial \bar u_2}+\bar C_3 \frac{\partial u_1}{\partial \bar u_3}$$
$$C_2 = \displaystyle \bar C_1 \frac{\partial u_2}{\partial \bar u_1}+\bar C_2 \frac{\partial u_2}{\partial \bar u_2}+\bar C_3 \frac{\partial u_2}{\partial \bar u_3}$$
$$C_3 = \displaystyle \bar C_1 \frac{\partial u_3}{\partial \bar u_1}+\bar C_2 \frac{\partial u_3}{\partial \bar u_2}+\bar C_3 \frac{\partial u_3}{\partial \bar u_3}$$
    or 
    <br>
$$C_p = \displaystyle \sum_{q=1}^{3}\ \ \bar C_q \frac{\partial u_p}{\partial \bar u_q} =  \bar C_q \frac{\partial u_p}{\partial \bar u_q}   \qquad (p=1,2,3)$$
and by interchanging coordinates and using the summation convention<br>
    $$\bar C_p = \displaystyle    C_q \frac{\partial \bar u_p}{\partial  u_q}   \qquad (p=1,2,3)$$
    <br> The $C_p$ and $\bar C_p$ are known as the components of a contravariant tensor of the first rank (or contravariant vector). The quantities  $C_1,\ C_2,\ C_3$ of $(u_1,\ u_2,\ u_3)$ are related to the quantites $\bar C_1,\ \bar C_2,\ \bar C_3$  of coordinates $(\bar u_1,\ \bar u_2,\ \bar u_3)$ via the transformation equations above, and this completes the definition.
    

<h1 style="color:red">Covariant Vector</h1>


For the covariant components of $\mathbf A$ the respective components in the systems $(u_1,\ u_2,\ u_3)$ and $(\bar u_1,\ \bar u_2,\ \bar u_3)$  are respectively <br>
$$\mathbf A = c_1 \nabla u_1 + c_2 \nabla u_2 +c_3 \nabla u_3= \bar c_1 \nabla \bar u_1 + \bar c_2 \nabla \bar u_2 +\bar c_3 \nabla \bar u_3$$

and since \(\bar u_p = \bar u_p (u_1,\ u_2,\ u_3)\) with p=1,2,3<br>
$$ \frac{\partial \bar u_p}{\partial x} = \frac{\partial \bar u_p }{\partial u_1}\frac{\partial u_1}{\partial x}+\frac{\partial \bar u_p }{\partial u_2 }\frac{\partial u_2}{\partial x}+\frac{\partial \bar u_p }{\partial u_3}\frac{\partial u_3}{\partial x}$$
    
$$ \frac{\partial \bar u_p}{\partial y} = \frac{\partial \bar u_p }{\partial u_1}\frac{\partial u_1}{\partial y}+\frac{\partial \bar u_p }{\partial u_2 }\frac{\partial u_2}{\partial y}+\frac{\partial \bar u_p }{\partial u_3}\frac{\partial u_3}{\partial y}$$
    
$$\frac{\partial \bar u_p}{\partial z} = \frac{\partial \bar u_p }{\partial u_1}\frac{\partial u_1}{\partial z}+\frac{\partial \bar u_p }{\partial u_2 }\frac{\partial u_2}{\partial z}+\frac{\partial \bar u_p }{\partial u_3}\frac{\partial u_3}{\partial z}$$
    
for p=1,2,3. These are equations (2). 
<br> also <br>
    $$\mathbf A = c_1 \nabla u_1 + c_2 \nabla u_2 +c_3 \nabla u_3 = \displaystyle \left( c_1 \frac{\partial u_1}{\partial x} +c_2 \frac{\partial u_2}{\partial x}+c_3 \frac{\partial u_3}{\partial x}\right)\mathbf i +\left( c_1 \frac{\partial u_1}{\partial y} +c_2 \frac{\partial u_2}{\partial y}+c_3 \frac{\partial u_3}{\partial y}\right)\mathbf j + \left( c_1 \frac{\partial u_1}{\partial z} +c_2 \frac{\partial u_2}{\partial z}+c_3 \frac{\partial u_3}{\partial z}\right)\mathbf k \qquad\qquad (3)$$
<br> and <br>
$$\bar c_1 \nabla \bar u_1 + \bar c_2 \nabla \bar u_2 +\bar c_3 \nabla \bar u_3 = \displaystyle \left( \bar c_1 \frac{\partial \bar u_1}{\partial x} +\bar c_2 \frac{\partial \bar u_2}{\partial x}+\bar c_3 \frac{\partial \bar u_3}{\partial x}\right)\mathbf i +\left( \bar c_1 \frac{\partial \bar u_1}{\partial y} +\bar c_2 \frac{\partial \bar u_2}{\partial y}+\bar c_3 \frac{\partial \bar u_3}{\partial y}\right)\mathbf j + \left( \bar c_1 \frac{\partial \bar u_1}{\partial z} +\bar c_2 \frac{\partial \bar u_2}{\partial z}+\bar c_3 \frac{\partial \bar u_3}{\partial z}\right)\mathbf k \qquad\qquad (4) $$
<br>
setting 

and equating coefficients of $\mathbf i, \mathbf j,\ \mathbf k$ gives : <br>
$$c_1 \frac{\partial u_1}{\partial x} +c_2 \frac{\partial u_2}{\partial x}+c_3 \frac{\partial u_3}{\partial x} = \bar c_1 \frac{\partial \bar u_1}{\partial x} +\bar c_2 \frac{\partial \bar u_2}{\partial x}+\bar c_3 \frac{\partial \bar u_3}{\partial x}$$
$$c_1 \frac{\partial u_1}{\partial y} +c_2 \frac{\partial u_2}{\partial y}+c_3 \frac{\partial u_3}{\partial y} =  \bar c_1 \frac{\partial \bar u_1}{\partial y} +\bar c_2 \frac{\partial \bar u_2}{\partial y}+\bar c_3 \frac{\partial \bar u_3}{\partial y}$$
$$c_1 \frac{\partial u_1}{\partial z} +c_2 \frac{\partial u_2}{\partial z}+c_3 \frac{\partial u_3}{\partial z} = \bar c_1 \frac{\partial \bar u_1}{\partial z} +\bar c_2 \frac{\partial \bar u_2}{\partial z}+\bar c_3 \frac{\partial \bar u_3}{\partial z}$$
<br>
Substituting equations (2) in to the first term ...<br>
    
$$c_1 \frac{\partial u_1}{\partial x} +c_2 \frac{\partial u_2}{\partial x}+c_3 \frac{\partial u_3}{\partial x} = \bar c_1 \left(\frac{\partial \bar u_1 }{\partial u_1}\frac{\partial u_1}{\partial x}+\frac{\partial \bar u_1 }{\partial u_2 }\frac{\partial u_2}{\partial x}+\frac{\partial \bar u_1 }{\partial u_3}\frac{\partial u_3}{\partial x} \right) +\bar c_2 \left(\frac{\partial \bar u_2 }{\partial u_1}\frac{\partial u_1}{\partial x}+\frac{\partial \bar u_2 }{\partial u_2 }\frac{\partial u_2}{\partial x}+\frac{\partial \bar u_2 }{\partial u_3}\frac{\partial u_3}{\partial x}\right)+\bar c_3 \left(\frac{\partial \bar u_3 }{\partial u_1}\frac{\partial u_1}{\partial x}+\frac{\partial \bar u_3 }{\partial u_2 }\frac{\partial u_2}{\partial x}+\frac{\partial \bar u_3 }{\partial u_3}\frac{\partial u_3}{\partial x}\right)$$
    
<br> expanding these out and factorizing to equate coefficients <br>
    
$$c_1 \frac{\partial u_1}{\partial x} +c_2 \frac{\partial u_2}{\partial x}+c_3 \frac{\partial u_3}{\partial x} = $$
    
$$\bar c_1 \frac{\partial \bar u_1 }{\partial u_1}\frac{\partial u_1}{\partial x}+\bar c_1\frac{\partial \bar u_1 }{\partial u_2 }\frac{\partial u_2}{\partial x}+\bar c_1\frac{\partial \bar u_1 }{\partial u_3}\frac{\partial u_3}{\partial x} + \bar c_2\frac{\partial \bar u_2 }{\partial u_1}\frac{\partial u_1}{\partial x}+\bar c_2\frac{\partial \bar u_2 }{\partial u_2 }\frac{\partial u_2}{\partial x}+\bar c_2\frac{\partial \bar u_2 }{\partial u_3}\frac{\partial u_3}{\partial x}+ \bar c_3\frac{\partial \bar u_3 }{\partial u_1}\frac{\partial u_1}{\partial x}+\bar c_3\frac{\partial \bar u_3 }{\partial u_2 }\frac{\partial u_2}{\partial x}+\bar c_3\frac{\partial \bar u_3 }{\partial u_3}\frac{\partial u_3}{\partial x} $$      
     
$$ = \left( \bar c_1 \frac{\partial \bar u_1 }{\partial u_1}+ \bar c_2\frac{\partial \bar u_2 }{\partial u_1} + \bar c_3\frac{\partial \bar u_3 }{\partial u_1}\right)\frac{\partial u_1}{\partial x}+\left(\bar c_1\frac{\partial \bar u_1 }{\partial u_2 }+\bar c_2\frac{\partial \bar u_2 }{\partial u_2 }+\bar c_3\frac{\partial \bar u_3 }{\partial u_2 }\right)\frac{\partial u_2}{\partial x}+\left(\bar c_1\frac{\partial \bar u_1 }{\partial u_3}+\bar c_2\frac{\partial \bar u_2 }{\partial u_3}+\bar c_3\frac{\partial \bar u_3 }{\partial u_3}\right)\frac{\partial u_3}{\partial x}$$       

<br> Therefore the following relations hold <br>
$$c_1 = \bar c_1 \frac{\partial \bar u_1 }{\partial u_1}+ \bar c_2\frac{\partial \bar u_2 }{\partial u_1} + \bar c_3\frac{\partial \bar u_3 }{\partial u_1}$$
$$c_2 = \bar c_1\frac{\partial \bar u_1 }{\partial u_2 }+\bar c_2\frac{\partial \bar u_2 }{\partial u_2 }+\bar c_3\frac{\partial \bar u_3 }{\partial u_2} $$
$$c_3 = \bar c_1\frac{\partial \bar u_1 }{\partial u_3}+\bar c_2\frac{\partial \bar u_2 }{\partial u_3}+\bar c_3\frac{\partial \bar u_3 }{\partial u_3}$$
    and this is usually written as <br>
$$c_p = \bar c_q \displaystyle \frac{\partial \bar u_q}{\partial u_p} \qquad\qquad (p=1,2,3)$$
    and similarly
$$\bar c_p =  c_q \displaystyle \frac{\partial  u_q}{\partial \bar u_p} \qquad\qquad (p=1,2,3)$$

using Einstein's summation convention over the repeated indices \(q\). 
<br> 
    so when the the quantities $c_1,\ c_2,\ c_3$ of a coordinate system $u_1,\ u_2,\ u_3$ are related to the quantities $\bar c_1,\ \bar c_2,\ \bar c_3 $ another system $\bar u_1,\ \bar u_2,\ \bar u_2$ according to the transformation equations above then the quantites are known as the components of a covariant tensor of the first rank (or covariant vector).
