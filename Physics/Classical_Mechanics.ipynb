## Kinematics in 3 dimensions
   
Average Velocity:     $\bar{ v}=\frac{\Delta \mathbf r}{\Delta t}$
   
Instantaneous\ Velocity:  $\mathbf v=\frac{d\mathbf r}{d t} $
   
Average Acceleration: $\bar {a}=\frac{\Delta \mathbf v}{\Delta t}$
   
Instantaneous Acceleration: $\mathbf a=\frac{d\mathbf v}{d t}= \frac{d^2\mathbf r}{d t^2} $
   
Motion with constant Acceleration :
   
$\mathbf v = \mathbf v_0 + \mathbf a t $
   ...and
      
$\mathbf r-\mathbf r_0 = \mathbf v_0 t + \frac{1}{2}\mathbf a t^2$
   
### Motion of a Projectile   

$v_x = v_{0_x}+v_0 \cos \theta \qquad v_z = v_{0_z}-g t = v_0 \sin \theta - gt$   
   
and
   $x = v_{0_z}t-\frac{1}{2}gt^2$   

### Range, maximum height, and time of flight (over flat ground)
   
Range $ x_{max} = \frac{v_0^2\sin {2 \theta}}{g}$
   
maximum height $z_{max} = \frac{v_0^2\sin^2 { \theta}}{2 g}$
   
time of flight $t_{flight} = \frac{2v_0\sin { \theta}}{g}$

Centripetal Acceleration in uniform circular motion\ \ $a = v^2/r$
   
### Galilean transformations $ t'=t$
   
position change $ \mathbf r'=\mathbf r - \mathbf V_0t $
   
velocity change $\mathbf v'=\mathbf r - \mathbf V_0 $


## Dynamics Newtons Laws  
---
   
#### Newtons First Law:
>In an inertial frame, a body at rest remains at rest and a body in motion continues to move at constant velocity unless acted on by an external force.
   
Definition of mass: $ m/m_s=a/a_s$
   
#### Newtons Second Law:
> $m\ \mathbf a=\mathbf F_{net} = \sum \mathbf F$
   
Superposition of forces: $\mathbf F_{net}=\mathbf F_1+\mathbf F_2+\mathbf F_3 +...$
   
#### Newton's Third Law: 
>Whenever a body exerts a force on another body, the latter exerts a force of equal magnitude and opposite direction on the former.
   
Momentum: $ \mathbf p = m\mathbf v $

and

$\frac{d \mathbf p}{d t} =\mathbf F$

---

Weight: $\text{magnitude} = m\ g$, direction = downward
    
Kinetic Friction: $f_k=m\ u_k\ N$ 
    
Static Friction: $f_s\leq\ m\ u_s\ N$
         
Restoring Force of a spring: $F=-k\ x$

##  Work and Energy

Work done by a constant force: 
   
$W=\mathbf F\cdot\Delta \mathbf r = F_x\Delta x+F_y \Delta y + F_z \Delta z$
      
Work done by a Variable Force: 
   
$W=\int_{p_1}^{p_2}\mathbf F.d\mathbf r=\int_{p_1}^{p_2}F_x\ dx+\int_{p_1}^{p_2}F_y\ dy + \int_{p_1}^{p_2}F_z\ dz$
    
Work done by Gravity: 
   
$W=-m\ g \Delta z$
      
Kinetic Energy: 
   
$K=\frac{1}{2}m v^2 = \frac{p^2}{2m} $ [because\ $P=m\mathbf v$ ]
    
Work Energy Theorem: 
   
$\Delta K = W$
     
Gravitational Potential Energy: 
   
$ m\ g\ z$

Mechanical Energy: 

$E=K+m\ g\ z$
   
Conservation of mechanical Energy:  

$E=K+P= K+m\ g\ z =[\text{const}] $

Conservative Force: The work done by the Force does not depend on the path; it depends only on the positions of the end points of the path.
   
Potential Energy of a Conservative Force: $U(p) = -\int_{p_0}^{p}\mathbf F.d\mathbf r + U(p_0)$
   
Mechanical Energy: $E=K+U$
   
Conservation Of Energy: $E=K+U=[\text{const}]$
   
Potential Energy of a Spring: $U(x) = \frac{1}{2}k\ x^2 $
   
Loss of Mechanical Energy by Non-Conservative force: $\Delta E = W_{_{nonconserv}}$

Force as a Derivative Of Potential Energy: $F_x = \frac{\partial \mathbf U}{\partial x},\quad F_y = \frac{\partial \mathbf U}{\partial y},\quad F_z = \frac{\partial \mathbf U}{\partial z}$

Force is the gradient of the potential: $\mathbf F = -\nabla \phi$

Mass is a form of Energy: $E=m c^2$

Energy has mass: $\Delta m=\frac{\Delta E}{c^2}$

Average Power:$\bar{ P}=\frac{\Delta W}{\Delta t}$

Instantaneous Power: $P=\frac{d W}{dt}$

Mechanical Power delivered by a force: $P=\mathbf F.\mathbf v$

##  Gravitation 
   
Law Of Universal Gravitation:  
   
$F=\frac{G\ M\ m}{r^2}, \quad \ G=6.67\times 10^{-11} \frac{N.m^2}{kg^2} $

Acceleration Of Gravity On Earth:$ g=\frac{G\ M_E}{R_E^2}$

Circular Orbit Around Sun: $ v^2=\frac{GM_S}{r}$

Keplers First Law: $The orbits of the planets are ellipses with the sun at one focus.

Kepler's Second Law: The radial line segement from the sun to the planet sweeps out equal areas in equal times.

Kepler's Third Law: The square of the period is proportional to the cube of the semimajor axis of the planetary orbit.

Gravitational Potential Energy:  $ U=-\frac{G\ M\ m}{r}$

Energy for Circular Orbit around the Sun: $U=-\frac{G M_{S}\ m}{2r}$

Escape Velocity for Earth: $ v=\sqrt{\frac{2GM_E}{R_E}} $

Rocket Equation:  $ v-v_0=u \ln \left( \frac{M_0}{M}\right)$

Gauss' law: $\int_s \mathbf g\cdot d\mathbf S = -4\pi\ GM$ 


##  Systems of Particles    

Momentum of a system of Particles:  $ \mathbf P = \mathbf P_1+ \mathbf P_2+ \mathbf P_3+...+ \mathbf P_n$
   
Rate of change of momentum:   $ \frac{d \mathbf P}{dt}=\mathbf F_{ext}$
   
Conservation of momentum (in the absence of external forces):  $ \mathbf P=[const]$
   
Center of Mass:   $ \mathbf r_{cm}=\frac{m_1\mathbf r_1+m_1\mathbf r_1+...+m_n\mathbf r_n}{M}$
   
and $ r_{cm}=\frac{1}{M}\int \mathbf r \rho\  dV$
   
Momentum of a system of Particles:   $\mathbf P= M \mathbf v_{cm}$
   
Motion of the center of Mass:   $ M\mathbf a_{cm}=\mathbf F_{ext}$
   
Kinetic energy of a system of Particles:  $K=K_{int}+\frac{1}{2}v_{cm}^2$
   


##  Rotational kinematics of a Rigid Body    
   
Average angular velocity:  $\bar\omega=\frac{\Delta\phi}{\Delta t}$
   
Instantaneous angular velocity:  $\omega=\frac{d\phi}{dt}$
   
Average angular acceleration:  $\bar\alpha=\frac{\Delta \omega}{\Delta t}$
   
Instantaneous angular acceleration:  $ \alpha=\frac{d\omega}{d t}=\frac{d^2\phi}{dt^2}$
   
Speed of particle on rotating body:  $v=R\ \omega$
   
Acceleration of particle on rotating body: $ \ a_{tan}=R\ \alpha, \ \ \ \ a_{cent}=R\ \omega^2$
   
Motion with constant angular acceleration: $ \omega=\omega_0+\alpha t$
   
   and  $$\phi-\phi_0=\omega_0t+\frac{1}{2}\alpha t^2$$
   
   and  $$\alpha(\phi-\phi_0)=\frac{1}{2}(\omega^2-\omega_0^2)$$
                  



##  Dynamics of a Rigid Body    

Moment of Inertia:  $ I=\sum_{i=1}^n m_iR_i^2, $
                   $ I= \int \rho R^2\ dV$
   
Kinetic Energy of Rotation: $ K=\frac{1}{2}I \omega^2$
   
Parallel Axis Theorem:  $I=I_{cm}+Md^2$
   
Perpedicular axis theorem (for a flat plate in the x-y plane)L:    $I=I_z+I_y$

Angular momentum of a particle:  $\mathbf L=\mathbf r \times \mathbf p =m \mathbf r \times \mathbf v$

Angular momentum for continuous distribution:  $\mathbf L=\int \mathbf r \times \mathbf v dm$
   
Angular momentum for uniform circular motion: \ $L=m\ r\ v$
   
Angular momentum of a rigid body rotating about z-axis: \ $L_z =I\omega$
   
Torque:   $ \mathbf\tau =\mathbf r\times\mathbf F$

External Torque on System:  $\mathbf \tau_{ext} = \frac{d\mathbf L}{dt} $ where $\mathbf L$ is angular momentum.
       
Equation of rotational motion (fixed axis or axis through the center of mass): $  I\alpha = \tau$
   
Work done by torque:  $ W=\int\tau_z\ d\phi$
   
Conservation of Energy: $ E=\frac{1}{2}I\omega^2+U=[\text{constant}]$
   
Power delivered by a torque: $ P=\tau_z\ \omega$
   
Conservation of angular momentum: $I\omega = [\text{constant}]$

Conservation of angular momentum: $\mathbf \tau_{ext}= 0, \quad \mathbf L=[\text{constant}]$
       
Rolling motion (without slipping):  The axis through the point of contact is instantaneously a fixed axis.
    
Precession of a gyroscope: $\omega_p=rMg/L$



##  Statics And Elasticity    

Static Equilibrium:  The sums of the external forces and of the external torques of a rigid body are zero. Gravity effectively acts on the center of mass.
   
Mechanical advantage of lever:  $\frac{F'}{F}=\frac{l}{l'}$
   
Elongation of elastic material:  $\frac{\Delta L}{L}=\frac{1}{Y}\frac{F}{A}$
   
Shear:  $\frac{\Delta x}{h}=\frac{1}{S}\frac{F}{A}$
   
Compression of Volume:  $\frac{\Delta V}{V}=\frac{1}{B}\frac{F}{A}$

Stress: $\sigma = \frac{F}{A} $ where $A$ is the cross sectional area. $\sigma = \mathbf F\cdot\mathbf n / |A|$

Strain: $\epsilon_L = \frac{\Delta L}{L} = \sigma / Y $

Youngs Modulus (modulus of elasticity): $Y =  \frac{\sigma}{\epsilon} $

##  Oscillations    
   
Simple harmonic motion: $x=A\cos(\omega t+\delta)$
   
Period: $T=\frac{2\pi}{\omega}$
   
Frequency: $\nu = \frac{1}{T}=\frac{\omega}{2\pi}$
   
Equation of motion of simple harmonic oscillator: $m \frac{d^2x}{d t^2}=-k x$
   
Angular frequency of simple harmonic oscillator: $\omega=\sqrt\frac{k}{m}$
   
Energy of simple harmonic oscillator: $E=\frac{1}{2}kA^2=\frac{1}{2}m\ v_{max}^2$
   
Angular frequency and period of simple pendulum:$\omega=\sqrt{\frac{g}{l}}, \ \ \ \ T=2\pi\sqrt{\frac{l}{g}}$
   
Angular frequency of physical pendulum:$\omega=\sqrt{\frac{Mgl}{I}}$
   
Angular frequency of torsional pendulum:$\omega = \sqrt{\frac{\kappa}{I}}$
   
Fractional Energy loss per period of damped oscillator:$\frac{\Delta E}{E}=-2\pi\frac{\gamma}{\omega}$
   
Q of damped oscillator: $Q=\frac{\omega}{\gamma}$


## Waves    
   
General wave function: $y=f(x+vt)$
   
Wavelength and Frequency: $ \lambda v = v$
   
Harmonic wave: $y=A\cos k(x+vt)=A \cos (k\ x +\omega t)$
   
Wave number:   $k=\frac{2\pi}{\lambda}$
   
Angular Frequency: $\omega = 2 \pi \nu$
   
Wave equation:     $\frac{\partial^2 y}{dt^2}=v^2\frac{\partial^2y}{d x^2}$
   
Speed of a wave on a string:   $v=\sqrt{\mathbf F/\mu}$
   
Power of a wave on a string:  $P \propto v\omega^2A^2$
   
Superposition principle for two or more waves:  The net instantaneous deformation is the sum of the individual deformations.
   
Constructive interference:  Waves meet crest to crest.
   
Destructive interference:  Waves meet crest to trough.
   
Beat frequency:  $v_{beat}=v_1-v_2$
   
Standing harmonic wave:  $y=A\cos kx\cos\omega t$
   
Node: Point of zero oscillation.
   
Antinode: Point of maximum oscillation.
   
Wavelengths of normal modes of string:  $\lambda =2L,L,\frac{2}{3}L,...$
   
Eigenfrequencies:  $ v=\frac{v}{2L},\frac{v}{L},\frac{3v}{2L},...$



##  Fluid Statics and Dynamics    

Continuity equation for incompressible fluid:   $vA=[\text{constant}]$
   
Definition of pressure: $ p=\frac{\Delta F}{\Delta A}$, normal force on plane surface $F = \int p dA$

Pascal's Principle: Pressure applied to an enclosed fluid is transmitted in the fluid undiminished to every portion of the fluid and to the walls of the container.

Density: Mass per unit volume $\rho = \frac{m}{V}$

Specific Gravity: $S=\frac{\rho}{\rho_{\text{water}}}$, and $\rho_{\text{water}}=1000 \frac{kg}{m^3}$
   
Atmospheric pressure (average):  $1\ \text{atm} = 760\ mmHg=1.0\times 10^5 N/m^2\ or\ Pa$
   
Hydrostatic pressure (for incompressible fluid):  $p-p_0=-g \rho z$
   
Archimedies Principle:  The bouyant force has the same magnitude as the weight of displaced fluid.

Conservation of mass: Net flow inward over surface = rate of increase within surface (no sources or sinks)

Continuity Equation: For two sections of a tube, $\rho_1 A_1 v_1 =\rho_2 A_2 v_2 $, for incompressible the density is unit. 
   
Bernoulli's equation: $ \frac{1}{2}\rho v^2 + \rho gz+p=[constant]$


##  The Idea Gas and Kinetic Theory    
   
Ideal gas law:  $ pV=nRT \quad R=8.31\ J/K$
   
and $ pV=NkT \quad k=1.38\times 10^{-23}\ J/K$
   
Avogadro's number: $ N_A= R/k=6.02\times 10^{23}\ molecules/mole$
   
Temperature scales:   
	Absolute:  T	
    Celcius: $ T_C=T-273.15$   
    Fahrenheit: $ T_v=\frac{9}{5}T_C+32$
   
Root mean square speed: $ v_{rms}=\sqrt{\frac{3kT}{m}}$ (for ideal gas)
   
Kinetic energy of ideal monoatomic gas:  $ K=\frac{3}{2}NkT$
   
Equipartition theorem:   Each translational or rotational component of the random thermal motion of a molecule has an average kinetic energy of $\frac{1}{2}kT.$ Thus for a diatomic gas (rotational and translational kinetic energy), $E\approx \frac{5}{2} NkT$
   
Mean free math:  $\lambda=\frac{1}{4\pi R^2(N/V)}$
   
Random walk: $ r=\sqrt n\ \lambda$


##  Sound and Other Wave Phenomena  

Relation between excess pressure $p_e$ and longitudinal displacement $\chi$ :
   
   $p_e=-B \frac{\partial \chi}{\partial x} $, where $B$ is the bulk modulus of the medium. Both $p_e$ and $\chi$ satisfy the wave equation.
   
Wave Speed: $v=\sqrt{\frac{B}{\rho_0}}$ where $\rho_0$ is the density of the medium in equilibrium.

Wave Speed in Gas: $v=\sqrt{\frac{B_{\text{adibiatic}}}{\rho_0}}=\sqrt{\frac{\gamma p_0}{\rho_0}}=\sqrt{\frac{\gamma k T}{m}}$ where $p_0$ is the undisturbed gas pressure, $\gamma$ is the specific heat ratio, $k$ is Boltzmanns constant and $m$ is the mass of a molecule of the gas. 

Intensity of Sine wave: $I = \frac{1}{\text{Area}} \left(\frac{\partial U}{\partial t}\right)_{avg}=\frac{1}{2}\omega^2 A^2 \rho v$ ... is the average over one perdiod rate of transfer of energy per unit cross sectional area. Units of $(W/m^2)$

Decrease of intensity: decreases with the inverse square of the distance to the source. (no absorption). 

Intensity (loudness): = $10 log_{10}\left(\frac{I}{10^{-12}}\right) (dB)$ where $I$ is in $(W/m^2)$

Decrease of intensity of sound wave with distance:   $ I_2=I_1 \frac{r_1^2}{r_2^2}$
   
Speed of sound in air: $  \ 331\ m/s\ $
   
Standing waves in tube open at one end: $ \ \lambda_1=4L,\ \ \frac{4}{3}L, \ \ \frac{4}{5}L $
   
Doppler Shift: stationary emitter: $ v'=v\left(1\pm\frac{V_R}{v}\right)$
and $ v'=\frac{v}{\left(1\ \mp \frac{V_E}{R} \right)}$
   
Doppler shift for small speed: $   \frac{\Delta v}{v}=\pm \frac{V_R}{v}\ or\ \pm \frac{V_E}{v}$

##  Thermodynamics    

First law of thermodynamics:   $\Delta E=\Delta Q-\Delta W$,     ($\Delta Q$ and $\Delta W$ are not exact differentials)
   
Efficiency of a heat engine:   $e=1-\frac{Q_2}{Q_1}$
   
Efficiency of a Carnot engine:  $ e=1-\frac{T_2}{T_1}$
   
Carnot's theorem:  The efficiency of any engine cannot exceed that of a Carnot engine; the efficiency of any reversible engine equals that of a Carnot engine.
   
Clausius' theorem: (reversible cycle) $\int\frac{dQ}{T}=0$
   
Definition of Entropy: $S(A)=\int_{A_0}^{A}\frac{dQ}{T} + S(A_0)$

Statistical Definition of Entropy: $S= k\ ln \Omega$ where $k$ is Boltzmann's constant, $\Omega$ is the set of distinct microscopic states.
   
Second law of thermodynamics:  An engine operating in a cycle cannot transform heat into work without some other effect on its environment; or an engine operating in a cycle cannot transfer heat from a cold resovoir to a hot resovoir without some other effect on its environement.

Second law of thermodynamics (restated): In any process the total entropy of a system and its surroundings increases or (in a reversible process) does not change. 
   
Clausius' theorem (irreversible cycle):  $\int\frac{dQ}{T}\leq 0$
   
Change of entropy in irreversible process: $ S(B)-S(A)\geq \int_B^A\frac{dQ}{T}$
   
Third Law of thermodynamics:  At absolute zero the entropy is zero.
