<h1>1) The Dot And Cross Product</h1>

of two vectors <b>A</b> and <b>B</b> is denoted by $\quad \mathbf{A}\cdot\mathbf{B}\quad $ is defined as the product of the magnitudes of <b>A</b> and <b>B</b> and the cosine of the angle between. 
In symbols $\quad \mathbf{A}\cdot\mathbf{B} = |\mathbf{A}||\mathbf{B}|\cos(\theta)\quad  (0 \leq \theta \leq \pi)\quad $ and
$\quad\mathbf{A}\cdot\mathbf{B} \quad$ is a scalar not a vector

<h3>Laws</h3>

<b>1)</b>   $ \quad\mathbf{A}\cdot\mathbf{B} = \mathbf{B}\cdot\mathbf{A}\quad   $     commutative law for dot products
<br>
<b>2)</b> $\quad\mathbf{A}\cdot(\mathbf{B}+\mathbf{C}) = \mathbf{A}\cdot\mathbf{B} + \mathbf{A}\cdot\mathbf{C}\quad $   Distributive Law 
<br>
<b>3)</b> $\quad(m(\mathbf{A}\cdot\mathbf{B}) = (m\mathbf{A})\cdot\mathbf{B} = \mathbf{A}\cdot(m\mathbf{B}) = (\mathbf{A}\cdot\mathbf{B})m \quad $     when m is a scalar
<br>
<b>4)</b> $ \mathbf{i}\cdot\mathbf{i} = \mathbf{j}\cdot\mathbf{j} = \mathbf{k}\cdot\mathbf{k} = 1    \quad\mathbf{i}\cdot\mathbf{j} = \mathbf{j}\cdot\mathbf{k} = \mathbf{k}\cdot\mathbf{i} = 0 $
<br>
<b>5)</b> if $\quad \mathbf{A}= A_1 \mathbf{i} + A_2 \mathbf{j} + A_3 \mathbf{k} ,\quad \mathbf{B} = B_1 \mathbf{i} + B_2 \mathbf{j} + B_3 \mathbf{k}\quad $ <br>then<br>
    $\qquad \mathbf{A}\cdot\mathbf{B} = A_1 B_1 + A_2 B_2 + A_3 B_3\\ 
    \qquad \mathbf{A}\cdot\mathbf{A} = A_1^2 + A_2^2 + A_3^2 $
<br>
<b>6)</b> if $\quad A.B = 0\quad $ and <b>A</b> and <b>B</b> are not null vectors then <b>A</b> and <b>B</b> are perpendicular

The Cross or Vector Product of <b>A</b> and <b>B</b> is a vector 

$$\mathbf{C} =\mathbf{A }\times \mathbf{B}. $$

The magnitude of \( \mathbf{A }\times \mathbf{B} \) is defined as the product of the magnitudes of <b>A</b> and <b>B</b> and the sin of the angle between them. The direction of <b>A</b> and <b>B</b> is perpendicular to the place of <b>A</b> and <b>B</b> such that <b>A, B, C </b> form a right handed system 

$$ \large  |\mathbf{A }\times \mathbf{B} |= |\mathbf{A}  | | \mathbf{B}  |\sin \theta \mathbf{u}  $$

<h2>Continutity and differentiability.</h2>



A scalar function $\quad \large \phi (u) \quad $ is called continuous at u if

$$\large \lim_{\delta u \to 0}(\phi(u+\delta u)-\phi (u))=\phi (u)$$

Equivalently $\quad \large \phi (u)\quad $ is continuous at u if for each positive number $\quad \large \epsilon\quad $ we can find a positive number $\quad\large \delta\quad $ such that

$$\large  |(\phi(u+\delta u)-\phi (u))  | < \epsilon \quad whenever\quad  | \delta u \ | < \delta$$

<br>

A vector function $\quad \mathbf{R}(u) = r_1(u) \mathbf{i} + r_2(u) \mathbf{j} + r_3(u)\mathbf{k} \quad $ is continuous at u if the three scalar functions 

$$\large r_1(u),\quad r_2(u),\quad r_3(u)$$

are continuous at u or if 

$$\large \lim_{\delta u \to 0}(\mathbf R(u+\delta u)-\mathbf R(u))=\mathbf R(u)$$

and equivalently if

$$\large \left |(\mathbf R(u+\delta u)-\mathbf R(u)) \right |<\epsilon \quad whenever\quad \left | \delta u \right |<\delta$$

 A scalar or vector function is called differentiable at order n if its nth derivative exists. A function which is differentiable is neccessarily continuous but the converse is not true.

<h1>4) Gradient, Divergence 

     And Curl</h1>
    

Vector differential operator $ \nabla $ is defined by 

$$\nabla=\frac{\partial }{\partial x} \mathbf i+\frac{\partial }{\partial y} \mathbf j+\frac{\partial }{\partial z} \mathbf k$$

This operator possesses properties analogous to ordinary vectors. It is useful in finding the gradient, divergence and curl.

<h2>The Gradient</h2>

If $ \phi(x,y,z) $ is defined and differentiable at each point $(x, y, z)$ in a certain region of space (differentiable scalar field) . The gradient of $\phi$ 



$$\nabla \phi=\mathbf{grad}\ \phi=\left (\frac{\partial }{\partial x} \mathbf i+\frac{\partial }{\partial y} \mathbf j+\frac{\partial }{\partial z} \mathbf k \right )\phi=\frac{\partial \phi }{\partial x} \mathbf i+\frac{\partial \phi}{\partial y} \mathbf j+\frac{\partial \phi}{\partial z} \mathbf k 
$$
<br>


$ \mathbf{grad}\ \phi $ defines a vector field.

<br>
<br>
The component of $ \mathbf{grad}\ \phi $ in the direction of a unit vector <b>a</b> is  $ \mathbf{grad}\ \phi\cdot\mathbf a $ or  $ \nabla \phi\cdot\mathbf a $ and is called the <b>directional derivative</b> of $ \phi $ in the direction of $ \mathbf{a}$. This is the rate of change of $\phi $ at $ (x, y, z) $ in the direction $\mathbf{a}$


<h2>The Divergence</h2>

let $\mathbf V = \mathbf V(x,y,z)=v_1\ \mathbf i+v_2\ \mathbf j +v_3\ \mathbf k$ be defined and differentiable at each point $(x, y, z)$ in a certain region of space (V is a differentiable vector field) Then
the divergence of V written $\nabla\cdot\mathbf V = \mathbf{div}\ \mathbf V$ is

$$
    \displaystyle\nabla\cdot\mathbf V = \mathbf{div}\ \mathbf V=\left (\frac{\partial }{\partial x} \mathbf i+\frac{\partial }{\partial y} \mathbf j+\frac{\partial }{\partial z} \mathbf k \right ).(v_1\ \mathbf i + v_2\ \mathbf j + v_3\ \mathbf k)\\=\frac{\partial v_1 }{\partial x} +\frac{\partial v_2}{\partial y} +\frac{\partial v_3}{\partial z} $$

<h2>The Curl</h2>

If  $ \mathbf V = \mathbf V(x,y,z)=v_1\ \mathbf i+v_2\ \mathbf j +v_3\ \mathbf k $ is a differentiable vector field then the curl or rotation at V is
   
$$\nabla \times \mathbf V = \mathbf{curl}\ \mathbf V\ or \ \mathbf{rot}\ \mathbf V \\=\left (\frac{\partial }{\partial x} \mathbf i+\frac{\partial }{\partial y} \mathbf j+\frac{\partial }{\partial z} \mathbf k \right )\times(v_1\ \mathbf i + v_2\ \mathbf j + v_3\ \mathbf k)\\ = \begin{vmatrix} \mathbf i&\mathbf j &\mathbf k \\ \frac{\partial }{\partial x}& \frac{\partial }{\partial y}& \frac{\partial }{\partial z}\\ v_1 & v_2 & v_3 \end{vmatrix}\\ =\begin{vmatrix} \frac{\partial }{\partial y}&\frac{\partial }{\partial z} \\ v_2& v_3 \end{vmatrix}\mathbf i-\begin{vmatrix} \frac{\partial }{\partial x}&\frac{\partial }{\partial z} \\ v_1& v_3 \end{vmatrix}\mathbf j+\begin{vmatrix} \frac{\partial }{\partial x}&\frac{\partial }{\partial y} \\ v_1& v_2 \end{vmatrix}\mathbf k\\=\left ( \frac{\partial v_3 }{\partial y}-\frac{\partial v_2 }{\partial z} \right )\mathbf i-\left ( \frac{\partial v_3 }{\partial x}-\frac{\partial v_1 }{\partial z} \right )\mathbf j+\left ( \frac{\partial v_2 }{\partial x}-\frac{\partial v_1 }{\partial y} \right )\mathbf k$$

$$\mathbf{A}(x,y,z) = A_1 \mathbf{i} + A_2 \mathbf{j} + A_3 \mathbf{k}$$ 
be a vector function of position defined and continuous on C. Then the integral of the tangenital component of A along C from P1 to P2 written as

$$\int_{P_1}^{P_2} \mathbf A\cdot d\mathbf r=\int_C \mathbf A\cdot d\mathbf r=\int_C \left (A_1dx+A_2dy+A_3dz \right )$$
Is an example of a line integral. If <b>A</b> is the force <b>F</b> on a particle moving along C this line integral represents the work done by the force. If C is a closed curve then the integral is represented by the closed curve line integral symbol.

$$\int_{P_1}^{P_2} \mathbf A\cdot d\mathbf r = \oint_{C}\mathbf A\cdot d\mathbf r$$
In fluid mechanics the integral is called the circulation of <b>A</b> about C where A represents the velocity of a fluid. Any integral which is evaluated along a curve is called a line integral. These integrals can also be defined in terms of the limits of sums as in elementary calculus.





<h2>Theorem: </h2>

If $\mathbf A = \nabla\phi$, everywhere in a region R of space, defined by $a_1 \leq x \leq a_2\ and\ b_1 \leq y \leq b_2, c_1 \leq z \leq c_2\quad $ where $\quad\phi(x,y,z) $ is single valued and has continuous derivatives in R then<br>

1) $\int_{P_1}^{P_2} \mathbf A\cdot d\mathbf r $ is independant of the path C in R joining P1 and P2    <br><br>
        

2) $\oint_{C}\mathbf A\cdot d\mathbf r=0 $ 
around any closed curve C in R <br><br>

In such cases <b>A</b> is called a conservative vector field and $\phi$ its scalar potential. A vector field <b>A</b> is conservative if and only if 

$$\nabla \times \mathbf A=0 \quad or \quad \mathbf{curl}\ \mathbf A =0$$


or equivalently $$\mathbf A = \nabla \phi = \mathbf {grad}\ \phi.$$ In such cases $$\mathbf A.d\mathbf r =A_1dx + A_2dy +A_3 dz = d\phi $$ an exact differential.


<h2>Surface Integrals</h2>


Let S be a two side surface and one side of S is arbitrarily the positive side of S (if S is closed then it is usually the outer side) A unit normal <b>n</b> to any point on the positive side of S is called a positive or outward drawn normal.

The differential of surface area dS has a vector d<b>s</b> whose magnitude is dS and direction is <b>n</b> at S. Then $$d\mathbf{s} =\mathbf{n}dS$$


$$
\iint \mathbf A\cdot d\mathbf s=\iint \mathbf A\cdot\mathbf n\ dS
$$

Is an example of a surface integral called the flux of <b>A</b> over S. Other surface integrals are 


$$
\iint \phi\ dS,\ \ \ \ \iint \phi\ \mathbf n\ dS,\ \ \ \ \iint \mathbf A \times d\mathbf s
$$


where $\phi$ is a scalar function. These integrals can also be defined in terms of limits of sums. There is a symbol for integration over a closed surface. 


To evaluate surface integrals it is convenient to express them as double integrals taken over the projected area of the surface S on one of the coordinate planes. This is possible if any line perpendicular to the coordinate plane chosen meets the surface in no more than one point. Usually S can be subdivided into surfaces that do satisfy the restriction.

$$
\iiint \mathbf A d V , \ \ \ \ \iiint \phi dV $$  are examples of Volume integrals or space integrals.

<h1>6) The Divergence Thereom, Stokes Thereom and related Integral Theorems</h1>
 
<h2>
The Divergence Theorem of Gauss 
 </h2>

states that if V is the volume bounded by a closed surface S and <b>A</b> is a vector function of postion with continuous derivatives then

$$
\iiint \mathbf{div}\ \mathbf A\ dV=\iiint \nabla.\mathbf A\ dV=\iint \mathbf A.\mathbf n dS=\iint \mathbf A.d\mathbf s 
$$
where <b>n</b> is the positive outward normal to S.

<h2>
Stokes Theorem
 </h2>   

states that if S is an open two sided surface bounded by a closed non-intersecting curve C (simple closed curve) then if A has continuous derivatives
$$
\
    \oint_C \mathbf A.d \mathbf r=\iint \mathbf {curl}\ \mathbf A.\mathbf n\ dS = \iint (\nabla \times \mathbf A) \cdot d\mathbf s
$$

where C is traversed in the positive direction. The direction of C is called positive if an observer walking on the boundary of S in this direction with head pointing in the direction of positive normal to S, has the surface on his left.


<h2>
Greens Theorem in the plane
</h2>

If R is a closed region of the xy plane bounded by a simple closed curve C and if M and N are continuous functions of x and y having continuous derivatives in R then

$$
\oint_C (M dx + N dy)=\iint_R \left( \frac{\partial N}{\partial x}-\frac{\partial M}{\partial y} \right )dx\ dy        
$$

where C is traversed in the positive (CCW) direction.  Greens integral in the plane is a special case of stokes theorem. Also Gauss divergence theorem is a generalization of Greens theorem where the plane region R and its closed boundary curve C are replaced by a space region V and a closed boundary surface S. 

    

Greens theorem also holds for regions bounded by a finite number of simple closed curves which do not intersect.

<h1>7) Curvilinear Coordinates</h1>

<h2>Transformation of coordinates:</h2> 

Let the rectangular coordinates $(x,\ y,\ z) $ of any point be expressed as functions of $(u_1, u_2, u_3) $ so that <br><br>


1) $ \qquad    x = x(u_1,\ u_2,\ u_3),\qquad     y = y(u_1,\ u_2,\ u_3),\qquad   z= z(u_1,\ u_2,\ u_3) $


<br><br>
suppose that 1) can be solved for $ (u_1, u_2, u_3) $ in terms of $ (x, y, z) $ <br><br>



2) $ \qquad    u_1 = u_1(x,\ y,\ z),\qquad      u_2 = u_2(x,\ y,\ z),\qquad     u_3 = u_3(x,\ y,\ z)$


<p>the functions in <b>1)</b> and <b>2)</b> are assumed to be single-valued and to have continuous derivatives so that the correspondance between $(x,\ y,\ z)$ and $(u_1,\ u_2,\ u_3)$is unique.</p>
<br>
<p>In practice this assumption may not apply at certain points and special consideration is required.</p>
<br>
<p>Given point P with rectangular coordinates $ (x,\ y,\ z) $ we can from <b>2)</b> associate a unique set of coordinates $(u_1,\ u_2,\ u_3) $ called the curvilinear coordinates of P. The sets of equations 1) and 2) define a transformation of coordinates.</p>

<h2>Orthogonal Curvilinear Coordinates</h2> 

<p>The surfaces $ u_1 = c_1$, $ u_2 = c_2 $ and $ u_3 = c_3 $ where$ c_1, c_2, c_3 $ are constants are called coordinate surfaces and each pair of these surfaces intersect in curves called coordinate curves or lines. If the coordinate system is called orthogonal. The $ (u_1,\ u_2,\ u_3) $  coordinate cutves of a curvilinear system are analogous to the $ (x,\ y,\ z) $ axes of a rectangular system.</p>



<h2>Unit Vectors in Curvilinear Systems</h2> 

Let $ \mathbf{r} = x \mathbf{i} + y \mathbf{j} + z \mathbf{k} $ be the position vector of a point P. Then <b>1)</b> can be written $ \mathbf{r}=r(u_1,u_2,u_3) $, A tangent vector to the $ u_1 $ curve at P (for which $ u_2 $ and $u_3 $ are constants) is $ \frac{dr}{du_1} $. Then a unit tangent vector in this direction is 


$$
\displaystyle\mathbf e_1 = \frac{\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_1} }{\left |\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_1} \right |} \ \ so\ that\ \frac{\mathrm{d}\mathbf r }{\mathrm{d} u_1}=h_1 \mathbf e_1 \ where \ \ h_1 = \left |\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_1} \right |  
$$


Similarly if $ \mathbf e_2 $ and $ \mathbf e_2 $ are unit tangent vectors to the $ u_2 $ and $ u_3 $ curves at P respectively then 


$$
\displaystyle\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_2}=h_2\ \mathbf e_2 \quad \ where \quad \ h_2 = \left |\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_2} \right |\qquad\qquad\displaystyle\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_3}=h_3\ \mathbf e_3 \quad \ where \quad \ h_3 = \left |\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_3} \right |
$$


The quantities $h_1,\ h_2,\ h_3$ are called scale factors. The unit vectors $ e_1,\ e_2,\ e_3 $ are in the  directions of increasing $ u_1,\ u_2,\ u_3 $ respectively.

Since $ \mathbf{grad} u_1 $ is a vector at P normal to the surface $ u_1 = c_1 $, a unit vector in this direction is given by



$$\mathbf{ E_1} = \frac{\mathbf{grad}\ u_1} {  | \mathbf{ grad }\ u_1  | }=\frac{\nabla u_1}{| \nabla u_1  |}$$



 Similarly 


 $$ \mathbf E_2 =\frac{\nabla u_2}{\left| \nabla u_2 \right |}  \\$$



 $$ \mathbf E_3 =\frac{\nabla u_3}{\left| \nabla u_3 \right |} \\$$



at P are normal to the surfaces $ u_2=c_2 $ and $ u_3=c_3 $ respectively.

Thus at each point P of a curvilinear coordinate system there exist in general two sets of unit vectors, $ \mathbf{e_1,\ e_2,\ e_3} $ tangent to the coordinate curves and $ \mathbf{E_1, E_2, E_3} $ normal to the coordinate surfaces. The sets become identical if and only if the curvilinear coordinate system is orthogonal. 

Both sets are analogous to the $ \mathbf{i,\ j,\ k} $ vectors in rectangular coordinates but are unlike them in that they may change directions from point to point. 

It can be shown that the sets 



$$\displaystyle\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_1},\ \frac{\mathrm{d}\mathbf r }{\mathrm{d} u_2},\ \frac{\mathrm{d}\mathbf r }{\mathrm{d} u_3}$$      and


$$
\nabla u_1, \ \nabla u_2,\ \nabla u_3
$$


constitute reciprocal systems of vectors. 
    

A vector can be represented in terms of the unit base vectors $ \mathbf{e_1,\ e_2,\ e_3} $ or $\mathbf{E_1,E_2,E_3} $ in this form


$$
A = A_1 \mathbf{e_1} + A_2 \mathbf{e_2} + A_3 \mathbf{e_3} = a_1 \mathbf{E_1} + a_2 \mathbf{E_2} + a_3 \mathbf{E_3} 
$$


where the $ A_i $ and $ a_i $are the respective components of <b>A</b> in each system.


We can also represent <b>A</b> in terms of the base vectors  


$$
\displaystyle\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_1},\ \frac{\mathrm{d}\mathbf r }{\mathrm{d} u_2},\ \frac{\mathrm{d}\mathbf r }{\mathrm{d} u_3}        \quad or \quad         \nabla u_1, \ \nabla u_2,\ \nabla u_3
$$


which are called unitary base vectors but are not unit vectors.




$$
\displaystyle\mathbf A = C_1\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_1} + C_2 \frac{\mathrm{d}\mathbf r }{\mathrm{d} u_2} + C_3 \frac{\mathrm{d}\mathbf r }{\mathrm{d} u_3} = C_1 \alpha_1 + C_2 \alpha_2 + C_3 \alpha_3
$$

and     


$$
\mathbf A =c_1 \nabla u_1 + c_2 \nabla u_2 + c_3 \nabla u_3 = c_1 \beta_1 + c_2 \beta_2 + c_3 \beta_3
$$

where $ C_1,\ C_2,\ C_3 $are called the <b>contravariant components</b> of <b>A</b> and $ c_1,\ c_2,\ c_3 $ are the <b>covariant components</b> of <b>A</b>. 


$$
\alpha_\rho = \frac{\partial \mathbf r}{\partial u_\rho}   ,\qquad          \beta_\rho = \nabla u_\rho,\qquad          \rho =1,\ 2,\ 3         
 $$ 


<h2>Arc Length and Volume Elements</h2>



From \( \mathbf r = \mathbf r(u_1,\ u_2,\ u_3) \) we have


$$\displaystyle d\mathbf r = \frac{\mathrm{d}\mathbf r }{\mathrm{d} u_1}d u_1 +\frac{\mathrm{d}\mathbf r }{\mathrm{d} u_2}d u_2 + \frac{\mathrm{d}\mathbf r }{\mathrm{d} u_3} d u_3=h_1 d u_1 \mathbf e_1+h_2 d u_2 \mathbf e_2+h_3 d u_3 \mathbf e_3$$

Then the differential arc length ds is determined from $ds_2 = d \mathbf{r}\cdot d \mathbf{r} $. For orthogonal systems,


$$\mathbf e_1.\mathbf e_2=\mathbf e_2.\mathbf e_3=\mathbf e_3.\mathbf e_1=0\quad   and\\ds^2=h_1^2d u_1^2+h_2^2d u_2^2+h_3^2d u_3^2$$

Along a $ u_1 $ curve, $ u_2 $ and $ u_3 $ are constants so that $ d \mathbf{r} = h_1 du_1 \mathbf{e_1} $. Then the differential arc length $ds_1 $ along $u_1 $ at P is $ h_1 du_1 $. Similarly the differential arc length along $ u_2 $ and $ u_3 $ at P are $ds_2 = h_2 du_2 $ and $ds_3 = h_3 du_3 $ <br>


    The volume element for an orthogonal curvilinear system is 





$$ d V=|(h_1 d u_1 \mathbf e_1).(h_2 d u_2 \mathbf e_2) \times h_3 d u_3 \mathbf e_3|=h_1 h_2 h_3\ d u_1 d u_2 d u_3\quad $$ since $$\quad |\mathbf e_1.\mathbf e_2 \times \mathbf e_3| = 1 $$

<h2>The Gradient, Divergence and Curl</h2>

can be expressed in terms of curvilinear coordinates. If $ \phi $ is a scalar function and 

$ \displaystyle \mathbf{A} = A_1 \mathbf{e_1} + A_2 \mathbf{e_2} + A_3 \mathbf{e_3} $ a vector function of orthogonal curvilinear coordinates $ u_1,\ u_2,\ u_3$ then the following is valid
<br>
<br>
1) $ \displaystyle \nabla \Phi = \mathbf{grad}\ \Phi = \frac{1}{h_1}\frac{d\Phi}{du_1}\mathbf e_1+ \frac{1}{h_2}\frac{d\Phi}{du_2}\mathbf e_2+ \frac{1}{h_3}\frac{d\Phi}{du_3}\mathbf e_3
$
<br>
<br>
2) $\displaystyle \nabla\cdot\mathbf A=\mathbf{div}\ \mathbf A=\frac{1}{h_1 h_2 h_3}\left [ \frac{\partial }{\partial u_1}(h_2 h_3 A_1)+\frac{\partial }{\partial u_2}(h_3 h_1 A_2)+\frac{\partial }{\partial u_3}(h_1 h_2 A_3) \right ]
$
<br>
<br>
3) $ \displaystyle \nabla \times \mathbf A = \mathbf{curl}\ \mathbf A = \frac{1}{h_1 h_2 h_3}\begin{vmatrix} h_1\mathbf e_1 & h_2\mathbf e_2 &h_3\mathbf e_3 \\ \frac{\partial }{\partial u_1} & \frac{\partial }{\partial u_2} & \frac{\partial }{\partial u_3} \\ h_1 A_1 &h_2 A_2 &h_3 A_3 \end{vmatrix} 
$
<br>
  <br>  
4) $ \displaystyle \nabla^2 \Phi = Laplacian\ of\ \Phi = \frac{1}{h_1 h_2 h_3}\left[ \frac{\partial }{\partial u_1}\left ( \frac{h_2 h_3}{h_1} \frac{\partial \Phi}{\partial u_1}\right ) + \frac{\partial }{\partial u_2}\left ( \frac{h_3 h_1}{h_2} \frac{\partial \Phi}{\partial u_2}\right ) + \frac{\partial }{\partial u_3}\left ( \frac{h_1 h_2}{h_3} \frac{\partial \Phi}{\partial u_3}\right ) \right]
$
<br>
<br>
if $ h_1 = h_2 = h_3 = 1\) and \(\mathbf e_1 \mathbf e_2 \mathbf e_3 $ are replaced by $\mathbf i \mathbf j \mathbf k$ these reduce to the usual expressions in rectangular coordinates where $ (u_1,\ u_2,\ u_3)$ are replaced by $(x, y, z) $.
