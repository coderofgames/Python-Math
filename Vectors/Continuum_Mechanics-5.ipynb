### Fourier's law of heat conduction
   
$$q_i = -k \frac{\partial \theta}{\partial x_i}$$
   
where $q_i$ is heat flux density
   
k is thermal conductivity
   
$\frac{\partial \theta}{\partial x_i}$ is the temperature gradient
   
the variable $k$ may be a function of temperature and direction, and in anisotropic materials it is represented by a second order tensor. 

### Balance of Energy
   
The second law of thermodynamics must also be satisfied. 
   
> A body of lower temperature cannot transfer heat to a body of higher tenperature without work.

$$\frac{d}{dt}\int_\lambda \left( e \rho + \frac{1}{2}\rho \dot{x}_i \dot{x}_i \right)d\Omega = \underbrace{\int_{\partial \lambda} \dot{x}_i T_{ij} n_j da}_{\text{internal stress}} + \underbrace{\int_{\lambda}\rho b_i \dot{x}_i d\Omega}_{\text{body force / gravity}} + \underbrace{\int_{\lambda}\rho r d\Omega}_{\text{ext radiation}} - \underbrace{\int_{\partial \lambda}q_i n_i da}_{\text{energy exchange through heat}}$$
   
In this case the energy of the l.h.s appears to be kinetic + potential energy, and the right hand side contains the internal stress energy from tractions, the energy due to gravbity, radiation and heat. The external radiation term is expressed as a volume integral ( I don't know why at this stage ?) and the two canidates for the divergence theorem are the stress tensor term and the heat exchange term, the heat exchange is negative if the body emits heat.

### Body at Rest
   
Kinetic energy is ignored
   
$$\frac{d}{dt}\int_\lambda  e \rho d\Omega =  \underbrace{\int_{\lambda}\rho r d\Omega}_{\text{ext radiation}} + \underbrace{\int_{\partial \lambda}q_i n_i da}_{\text{energy exchange through heat}}$$

### Body with Kinetic Energy
   
Differentiating the r.h.s wrt time and apply the divergence theorem and recognizing the volume is abitrary therefore if the integral is zero the integrand must be identically zero
   
$$ \dot{e}\rho + \rho \dot{x}_i \ddot{x}_i  =  \frac{\partial }{\partial x_j}(\dot{x}_i T_{ij}) + \rho b_i \dot{x}_i + \rho r - \frac{\partial q_i}{\partial dx_i} $$

and the term 
   
$$\displaystyle\begin{array}{rcl}  
&      \frac{\partial }{\partial x_j}(\dot{x}_i T_{ij}) & = & \frac{\partial \dot{x}_i}{\partial x_j}T_{ij}+ \dot{x}_i\frac{\partial T_{ij}}{\partial x_j} \\
&       & = & L_{ij}T_{ij}+ \dot{x}_i\frac{\partial T_{ij}}{\partial x_j} \\
&       & = &(D_{ij}+W_{ij})T_{ij}+ \dot{x}_i\frac{\partial T_{ij}}{\partial x_j} \\
&       & = &D_{ij}T_{ij}+ \dot{x}_i\frac{\partial T_{ij}}{\partial x_j} 
\end{array}$$
   
since the product of a skew symmetric and a symmetric matrix are zero.

We then factorize the terms from the conservation of linear momentum and 
   
$$\rho\dot{e} = T_{ij}D_{ij}+\rho r -\frac{\partial q_i}{\partial x_i}$$
   
if the body is rigid then strain rate $D_{ij}$ is zero
   
$$\rho\dot{e}=\rho r-q_{i,i}$$
   
and this implies internal (not potential?) $e$ is related only to the change in temperature
   
$$e=c(\theta-\theta_0) \implies \dot{e}= c\dot{\theta}$$
   
Then substituting in Fouriers equation for the $q_i$ we have
   
$$ \rho c \dot{\theta}=\rho r + k\frac{\partial^2 \theta}{\partial x_i \partial x_i} $$
   
where $c$ is the specific heat. This represents heat conduction in a rigid solid.
   

   



### Balance of Energy in the Reference Configuration
   
The term involving heat condution $\int_{\partial \lambda} q_i n_i da=\int_{\partial \lambda} q_i da_i$ can be rewritten using the rule of transformation of area
   
$$ da_i = J (F^{-1})_{Li}dA_L$$
   
then 
   
$$ Q_L = q_i J (F^{-1})_{Li}$$
   
and the term involving $T_{ij}$ is replaced with the appropriate term for $\tilde{T}_{iL}$. And after applying the divergence theorem and recognizing that the volume is arbitrary means that the if the integral is zero then the integrand is identically zero and factoring the conservation of linear momentum
   
$$ \rho_0 \dot{e}= \tilde{T}_{iL}\dot{x}_{i,L} + \rho_0 r - Q_{L,L}$$
   
### Entropy
   
   $$\frac{d}{dt}\int_{\lambda}\rho \underbrace{\eta}_{\text{entropy}} d\Omega \geq \underbrace{\int_\lambda \rho s d\Omega}_{\text{supply of entropy}} + \underbrace{\int_{\partial \lambda} h da}_{\text{contact entropy}}$$
Heat flux at a point depends on the normal. 
   
$$h = \frac{\mathbf q\cdot \mathbf n}{\theta_{\text{abs temp}}}$$
   
where $\theta > 0$, and zero degrees Kelvin is assumed. The entropy supply is
   
$$ s = \frac{r}{\theta} = \frac{\text{heat flux radiation}}{\text{temp (absolute)}}$$
   
We then obtain the Clausius Durhem inequality
   
 $$\frac{d}{dt}\int_{\lambda}\rho \underbrace{\eta}_{\text{entropy}} d\Omega \geq \underbrace{\int_\lambda \rho \frac{r}{\theta} d\Omega}_{\text{supply of entropy}} - \underbrace{\int_{\partial \lambda} \frac{ q_i  n_i}{\theta} da}_{\text{contact entropy}}$$
    
 and this is simplified thanks to the abitrary volume
    
 $$ \rho \dot{\eta} -\rho \frac{r}{\theta}  +\frac{\partial }{\partial x_i}\left(\frac{q_i}{\theta}\right)\geq 0 $$
    
or 

 $$ \rho \dot{\eta} -\rho \frac{r}{\theta}  +\frac{1}{\theta}\frac{\partial q_i}{\partial x_i}-\frac{\partial \theta}{\partial x_i}\frac{q_i}{\theta^2}\geq 0 $$
   
The system is in equilibrium if the equality holds. 

### Reduced Entropy Inequality
   
Since $\theta$ is positive the entropy inequality can be mulitplied throughout. Then subtracting the balance of energy gives
   
$$ \rho(\theta \dot{\eta}-\dot{e})+T_{ij}D_{ij} -\frac{q_i}{\theta}\frac{\partial \theta}{\partial x_i}\geq 0$$
   
### Helmholtz Free energy
   
This is defined as 
   
$$ \phi = e - \theta\eta$$
   
and since $\dot{\phi} = \dot{e} - \dot{\theta}\eta - \theta\dot{\eta}$ 

### Reduced entropy inequality in terms of Helmholtz free energy
   
$$ \rho(\dot{\phi}+\dot{\theta}\eta)+T_{ij}D_{ij} -\frac{q_i}{\theta}\frac{\partial \theta}{\partial x_i}\geq 0$$ 
   
This equation is said to critically define the material.+
 

### Reduced entropy inequality in the reference configuration
   
$$ \rho_0(\dot{\phi}+\dot{\theta}\eta)-\tilde{T}_{iL}\dot{F}_{iL} +\frac{Q_L}{\theta}\frac{\partial \theta}{\partial X_L}\leq 0$$ 
   
This equation is said to critically define the material.+
 

### Case with no rates of Change
   
In the reduced entropy inequality, if there are no rates of change
   
$$ \frac{q_i }{\theta} \frac{\partial \theta}{\partial x_i}\leq 0$$
   
this implies the angle between flux and temperature gradient must be greater than or equal to $90^\circ$. Substituting in Fouriers law (noting that again we use the negative sign in $q_i=-k \theta_{,i}$ to transform the direction of the inequality).
   
$$ \frac{k }{\theta}\frac{\partial \theta}{\partial x_i} \frac{\partial \theta}{\partial x_i}\geq 0$$
   
This implies $k \geq 0 $ since the magnitude of $\frac{\partial \theta}{\partial x_i} = \partial_i\theta = \theta_{,i}$ squared must be positive.

### Summary
   
To solve for the deformation of a material in Continuum mechanics, the following physical laws must hold
   
| Physical Law | Reference Configuration | Present Configuration |
|--------------|-------------------------|-----------------------|
| Balance Of Mass | $\rho_0 = J \rho $   | $\partial_t \rho+\partial_i (\rho \dot{x}_i) = 0 $|
| Conservation of Linear Momentum | $\rho_0 \ddot{x}_i = \rho_0 b_i+ \partial_L \tilde{T}_{iL}  = 0 $| $\rho \ddot{x}_i = \rho b_i+ \partial_j {T}_{ij}  = 0 $|
| Moment of Momentum | $ \tilde{T}_{iL }F_{jL} = \tilde{T}_{jL }F_{iL}$| $T_{ij} = T_{ji}$ |
| Balance of internal Energy |$ \rho_0 \dot{e}= \tilde{T}_{iL}\dot{x}_{i,L} + \rho_0 r - \partial_L Q_{L}$|$\rho\dot{e} = T_{ij}D_{ij}+\rho r -\partial_i q_i$|
| Entropy Inequality| $ \rho_0(\dot{\phi}+\dot{\theta}\eta)-\tilde{T}_{iL}\dot{F}_{iL} +\frac{Q_L}{\theta} \partial_L \theta\leq 0$ |$ \rho(\dot{\phi}+\dot{\theta}\eta)+T_{ij}D_{ij} -\frac{q_i}{\theta}\partial_i \theta\geq 0$|

   
All that is needed is **constituitive** equations that define the *constitution* of the material, intial conditions that define the initial state of the system, and boundary conditions that define the boundaries. A solution of these equations is called a thermodynamic process.

### Form of Constituitive Equations
   
If we define the temperature gradient $G=\partial_L \theta$ and note $\mathbf F = \mathbf F(\mathbf X, t)$ the deformation gradient is a function of the coordinates and time, and $\theta=\theta(\mathbf X,t)$ and $\mathbf G = \mathbf G(\mathbf X,t)$. Then to help solve for unknowns we define 
   
$$\displaystyle\begin{array}{rcl}  
\text{Stress Tensor} &  \tilde{T}_{iL}(\mathbf X,t) & = & \tilde{T}_{iL}( \mathbf F,\mathbf{\theta},\mathbf G)\\
\text{Heat Flux}&  \tilde{Q}_{L}(\mathbf X,t) & = & Q_{L}( \mathbf F,\mathbf{\theta},\mathbf G)\\
\text{Entropy}&  \eta (\mathbf X,t) & = & \eta( \mathbf F,\mathbf{\theta},\mathbf G)\\
\text{Internal Energy}&  e(\mathbf X,t) & = &e( \mathbf F,\mathbf{\theta},\mathbf G)\\
\text{Helmholtz Free Energy}&  \phi(\mathbf X,t) & = & ( \mathbf F,\mathbf{\theta},\mathbf G)
\end{array}$$
   
are all functions of $( \mathbf F,\mathbf{\theta},\mathbf G)$ then using the entropy inequality that must be satisfied for all thermodynamic processes, the derivatives are put into differential form
   
$$\dot{\phi} = \partial_t \phi = \frac{\partial \phi}{\partial F_{iL}}\dot{F}_{iL}  +\frac{\partial \phi}{\partial \theta}\dot{\theta} + \frac{\partial \phi}{\partial G_{L}}\dot{G}_{L}$$
   
and the reduced Entropy inequality  
   
$$ \rho_0(\dot{\phi}+\dot{\theta}\eta)-\tilde{T}_{iL}\dot{F}_{iL} +\frac{Q_L}{\theta} \partial_L \theta\leq 0$$
  
becomes (substiting our differential and $G_L = \partial_L \theta$)
   
$$ \rho_0\left(\frac{\partial \phi}{\partial F_{iL}}\dot{F}_{iL}  +\frac{\partial \phi}{\partial \theta}\dot{\theta} + \frac{\partial \phi}{\partial G_{L}}\dot{G}_{L}\right)+\rho_0\dot{\theta}\eta-\tilde{T}_{iL}\dot{F}_{iL} +\frac{Q_L}{\theta} G_L\leq 0$$
   
factorising
   
$$ \left(\rho_0\frac{\partial \phi}{\partial F_{iL}}-\tilde{T}_{iL} \right)\dot{F}_{iL}  +\left( \frac{\partial \phi}{\partial \theta} +  \eta\right)\rho_0 \dot{\theta}+ \rho_0\left(\frac{\partial \phi}{\partial G_{L}}\right)\dot{G}_{L} +\frac{Q_L}{\theta} G_L\leq 0$$

Now noting that we can fix $F, \theta, G$, choosing any value to make the equality hold
   
$$ \tilde{T}_{iL} = \rho_0\frac{\partial \phi}{\partial F_{iL}}, \quad \eta = -\frac{\partial \phi}{\partial \theta}, \quad \frac{\partial \phi}{\partial G_L} = 0$$
   
and this implies that Helmholtz free energy is not dependant on temperature gradient, so $\phi=\phi(\mathbf F,\theta)$ and since we define the Piola-Kirchoiff stress tensor $\tilde{T}_{iL}$ to a function of the derivative of the Helmholtz free energy $ \tilde{T}_{iL} = \rho_0\frac{\partial \phi}{\partial F_{iL}}$ then it cannot be a function of the temperature gradient.
   
Next we define a value
   
$$\alpha = Q_L \partial_L \theta = Q_L G_L \leq 0 $$
   
and to fix this for the equality to hold in the entropy inequality, then at the point where $G_L=0$ the derivative
   
$$\frac{\partial \alpha}{\partial G_L} = 0 \implies Q_L = 0$$
   
Therefore
   
$$ \mathbf Q_L (\mathbf F, \theta, 0) = 0$$

### Theorem (mine)
   
Let the unknowns in the continuum equations of the physical laws be coordinates of some state, phase or parameter space then the solutions to the continuum equations are the intersections of the surfaces in higher dimensional space, its curvature and compatibility relations can then be evaluated using generalized Gauss Wiengarten equations or intrinsic equaltions of the N-dimensional (18 in this case) dimensional space that they occupy. The discriminant of the second fundamental form will express the generalized osculating N-paraboloid at any point in the space.
