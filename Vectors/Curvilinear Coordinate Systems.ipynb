### 1. Cylindrical Coordinates $(\rho, \phi, z)$ 


$$
x =\rho \cos \phi,\quad y=\rho\sin\phi,\quad z=z
$$
where


$$
\rho\geq 0,\qquad 0\leq\phi<2\pi,\qquad -\infty < z < \infty
$$
   
   
$$ h_{\rho}=1,\ \ h_{\phi} =\rho,\ \ h_z=1
$$

### 2. Spherical Coordinates  $(r, \theta, \phi)$

$$
x =r \sin \theta \cos\phi,\quad y=r\sin\theta\sin\phi,\quad z=r\cos\theta
$$
where


$$
r\geq 0,\qquad 0\leq\phi<2\pi,\qquad  0\leq\theta\leq\pi$$
    
$$h_r=1,\qquad h_\theta =r,\qquad h_\phi=r\sin\theta
$$

```python
fig = plt.figure()
fig = plt.figure(figsize=(12, 10))
#ax = fig.gca(projection='3d')
theta = np.linspace(0, 2* np.pi, 100)

    
phi = np.linspace(0, np.pi, 100)


r = 2
ax = fig.add_subplot(2, 2, 1, projection='3d')
for j in range(0,6):
    phi_ = np.pi*j / 6
    z = r * np.cos(phi_)

    x = r * np.sin(phi_)* np.cos(theta)
    y = r * np.sin(phi_)* np.sin(theta)
    ax.plot(x, y, z, label='parametric curve')
    
ax = fig.add_subplot(2, 2, 2, projection='3d')
r = np.linspace(0,2,40)
for j in range(0,6):
    phi_ = np.pi / 2
    theta_ = 2*np.pi *j/6
    z = r * np.cos(phi_)


    x = r * np.sin(phi_)* np.cos(theta_)
    y = r * np.sin(phi_)* np.sin(theta_)
    
    ax.plot(x, y, z, label='parametric curve')
    ax.set_zlim(-1,1)

ax = fig.add_subplot(2, 2, 3, projection='3d')
r = 2
for j in range(0,12):
    
    theta_ = 2*np.pi *j/12
    z = r * np.cos(phi)


    x = r * np.sin(phi)* np.cos(theta_)
    y = r * np.sin(phi)* np.sin(theta_)
    
    ax.plot(x, y, z, label='parametric curve')
    
```

### 3. Parabolic Cylindrical Coordinates $(u,\ v,\ z)$

$$
x =\frac{1}{2}(u^2-v^2),\quad y=u\ v,\quad z=z
$$
where


$$
-\infty < u < \infty,\qquad v\geq 0,\qquad -\infty < z < \infty$$
   
$$
h_u=h_v = \sqrt{u^2+v^2},\qquad h_z=1
$$

```python

fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')
theta = np.linspace(0, 2* np.pi, 100)
u = np.linspace(-5, 5, 60)
for i in range(0,20):
    v = 4*i/20
    x  = 0.5 * (u**2 -v**2)
    y = u*v
    z = 0
    ax.plot(x, y, z, label='parametric curve')
    ax.set_xlim(-1,1)
    ax.set_ylim(-1,1)

v= np.linspace(-5, 5, 60)
for i in range(0,20):
    u = 4*i/20
    x  = 0.5 * (u**2 -v**2)
    y = u*v
    z = 0
    ax.plot(x, y, z, label='parametric curve')
    ax.set_xlim(-1,1)
    ax.set_ylim(-1,1)
```

### 4. Paraboloidal Coordinates  $(u,\ v,\ \phi)$


$$
x =u\ v\cos\phi,\quad y=u\ v \sin\phi,\quad z=\frac{1}{2}(u^2-v^2) 
$$
where


$$
u\geq 0,\qquad v\geq 0,\qquad 0 \leq\ phi < 2\pi$$

$$
h_u=h_v = \sqrt{u^2+v^2},\qquad h_\phi=u\ v
$$

```python
fig = plt.figure(figsize=(12, 10))

#ax = fig.gca(projection='3d')
phi = np.linspace(0, 2* np.pi, 100)
u = np.linspace(-5, 5, 60)

ax = fig.add_subplot(2, 2, 1, projection='3d')
for i in range(0,5):
    
    for j in range(0,5):
        u = j/5
        v = i/5
        x  = u*v * np.cos(phi)
        y = u*v * np.sin(phi)
        z = 0.5 *(u**2 - v**2)
        ax.plot(x, y, z, label='varying u and v')
    
ax = fig.add_subplot(2, 2, 2, projection='3d')
u = np.linspace(-1, 1, 60)
for i in range(0,5):
    v = (i)/5
    x  = u*v * np.cos(0)
    y = u*v * np.sin(0)
    z = 0.5 *(u**2 - v**2)
    ax.plot(x, y, z, label='parametric curve')
    
#ax = fig.add_subplot(2, 2, 3, projection='3d')
v = np.linspace(-1, 1, 60)
for i in range(0,5):
    u = (i)/5
    x  = u*v * np.cos(0)
    y = u*v * np.sin(0)
    z = 0.5 *(u**2 - v**2)
    ax.plot(x, y, z, label='parametric curve')
    
```

### 5. Elliptic Cylindrical Coordinates $(u,\ v,\ z)$


$$
x =a \cosh u \cos v,\quad y=a\sinh u\sin v,\quad z=z
$$
where


$$
 u\geq 0,\qquad 0\leq v<2\pi,\qquad -\infty < z < \infty $$
      
$$
 h_u=h_v = a\sqrt{\sinh^2u+\sin^2 v},\ \ h_z=1
$$


```python
fig = plt.figure(figsize=(12, 10))
#ax = fig.gca(projection='3d')
ax = fig.add_subplot(111, projection='3d')
u = np.linspace(0, 2, 100)



#ax = fig.add_subplot(2, 2, 1, projection='3d')
a = 1
for j in range(0,12):
    v = 2*np.pi*j /12

    x = a * np.cosh(u)*np.cos(v) 
    y = a * np.sinh(u) * np.sin(v)
    z=0

    ax.plot(x, y, z, label='parametric curve')
    
#ax = fig.add_subplot(2, 2, 2, projection='3d')

v = np.linspace(0, 2* np.pi, 100)
for j in range(0,6):
    u = j * 2 / 6
    x = a * np.cosh(u)*np.cos(v) 
    y = a * np.sinh(u) * np.sin(v)
    z=0

    ax.plot(x, y, z, label='parametric curve')
    ax.set_xlim(-2,2)
    ax.set_ylim(-2,2)
```

### 6. Prolate Spheroidal Coordinates $(\xi ,\ \eta,\ \phi)$


$$
x =a \sinh \xi \sin \eta \cos\phi,\quad y=a\sinh \xi \sin \eta \sin \phi,\quad z=a \cosh \xi cos \eta
$$
where 


$$
\xi\geq 0,\qquad 0\leq \eta\leq2\pi,\qquad 0\leq \phi<2\pi$$
   
$$h_\xi=h_\eta = a\sqrt{\sinh^2\xi+\sin^2 \eta},\qquad h_\phi=a \sinh \xi sin\eta
$$


```python
fig = plt.figure(figsize=(12, 10))

#ax = fig.gca(projection='3d')
eta = np.linspace(0, 2* np.pi, 100)
phi = np.linspace(0, 2* np.pi, 100)

#ax = fig.add_subplot(111, projection='3d')
a=1

    
ax = fig.add_subplot(2, 2, 1, projection='3d')

for i in range(0,6):
    phi_ = 0
    zeta = 2*np.pi * i / 6
    x  = a*np.sinh(zeta) * np.sin(eta) * np.cos(phi_)
    y =  a*np.sinh(zeta) * np.sin(eta) * np.sin(phi_)
    z = a*np.cosh(zeta) * np.cos(eta)
    ax.plot(x, y, z, label='parametric curve')
    
zeta = np.linspace(0, 5, 100)
ax = fig.add_subplot(2, 2, 2, projection='3d')
for i in range(0,6):
    phi_ = 0
    eta = 2*np.pi * i / 6
    x  = a*np.sinh(zeta) * np.sin(eta) * np.cos(phi_)
    y =  a*np.sinh(zeta) * np.sin(eta) * np.sin(phi_)
    z = a*np.cosh(zeta) * np.cos(eta)
    ax.plot(x, y, z, label='parametric curve')
    
eta = np.linspace(0, 2* np.pi, 100)
ax = fig.add_subplot(2, 2, 3, projection='3d')
for i in range(0,12):
    phi_ = 2*np.pi * i / 12

    x  = a*np.sinh(zeta) * np.sin(eta) * np.cos(phi_)
    y =  a*np.sinh(zeta) * np.sin(eta) * np.sin(phi_)
    z = a*np.cosh(zeta) * np.cos(eta)
    ax.plot(x, y, z, label='parametric curve')
#ax = fig.add_subplot(2, 2, 3, projection='3d')
```

<h2>8. Ellipsoidal Coordinates $$(\lambda, \mu, \nu)$$</h2>

Thus:
$$
\frac{x^2}{a^2-\lambda}+\frac{y^2}{b^2-\lambda}+\frac{z^2}{c^2-\lambda}=1,\ \ \ \ \lambda < c^2 < b^2 < a^2$$
and
$$
\frac{x^2}{a^2-\mu}+\frac{y^2}{b^2-\mu}+\frac{z^2}{c^2-\mu}=1,\ \ \ \ c^2 < \mu < b^2 < a^2
$$
and
$$
\frac{x^2}{a^2-\nu}+\frac{y^2}{b^2-\nu}+\frac{z^2}{c^2-\nu}=1,\ \ \ \ c^2 < b^2 < \nu < a^2
$$
and
$$h_\lambda = \frac{1}{2}\sqrt{\frac{(\mu-\lambda)(\nu-\lambda)}{(a^2-\lambda)(b^2-\lambda)(c^2-\lambda)}},       h_\mu= \frac{1}{2}\sqrt{\frac{(\nu-\mu)(\lambda-\mu)}{(a^2-\mu)(b^2-\mu)(c^2-\mu)}},     \\ \ \\
$$
and
$$h_\nu= \frac{1}{2}\sqrt{\frac{(\mu-\nu)(\lambda-\nu)}{(a^2-\nu)(b^2-\nu)(c^2-\nu)}}$$
