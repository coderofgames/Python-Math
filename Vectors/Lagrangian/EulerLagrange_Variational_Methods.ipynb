### Euler Lagrange Equation

The problem is to find the minimum curve between two points on a plane. What follows here was first derived by a Mathmagician. The equation of the curve is stated abstractly as a functional, meaning a function of $x$ and $\dot x$, perhaps $\ddot x$ (and often t). The function $f(x, \dot x)$ is continuously differentiable or "smooth". The parameters $x$ and $\dot x$ are both functions of $t$ such that $x = x(t)$ and $\dot x = \dot x(t)$

       
$$ J(f) = \int_a^b f(x,\ \dot x) dt $$
    
now if we introduce an error function 
    
$$ \displaystyle \eta(t)=\frac{x_1(t)-x(t)}{\varepsilon}$$

<img src="https://drive.google.com/uc?export=view&id=0B-Q-Pf4fc8LdSEZKTlpqZ2o0bUE"></img>

The $x$'s are curves and the function space is some kind of force field or potential field that has an effect on the position and velocity with respect to time. The difference between our actual value of $x(t)$ and our desired optimal value $x_1(t)$ is divided by a variable $\varepsilon$ that we seek to minimize. The difference between the two functions must be differentiable with respect to $\varepsilon$. 

    
$$ \frac{d}{d\varepsilon} (\eta(t,\varepsilon)) =    \frac{d}{d\varepsilon}\left ( \frac{x_1(t) - x(t)}{\varepsilon} \right ) $$ 
    
is defined, and continuously differentiable. Now the next stage is to express \(f(x, \dot x)\) in terms of $f(x_1,\dot x_1)$ and $\varepsilon$. 

$$ x_1(t) = x(t) + \varepsilon \eta(t) $$ 

and the change in $x_1$ with respect to value $\varepsilon$ is 

$$ \frac{\partial x_1(t, \varepsilon )}{\partial \varepsilon} = \eta(t) $$

now at the points a and b, the error curve $\eta(t)$ equals zero because at $t=a$ and $t=b$ the curve $x_1(t)$ has the same value as $x(t)$, since $\eta(t)$ defines the error.
    
    
now to add a small variation to $J(f)$, expressed as $\delta J(f) = J(\delta f) $ the infinitesimal operator passes inside the integral 
   
$$ \displaystyle \delta J(f) = \delta \int_a^b f(x,\ \dot x) dt = \int_a^b \delta f( x,\ \dot x) dt $$ 

Now the value $\delta f(x, \dot x) $ is the total differential with respect to the infinitesimal $\varepsilon$, the variable $x$ becomes  (or is substituted with) $x_1$ 
   
    
$$ \frac{\partial f}{\partial \varepsilon} = \frac{\partial f}{ \partial x_1} \frac{\partial x_1}{\partial \varepsilon} + \frac{ \partial f}{ \partial \dot x_1}\frac{\partial \dot x_1}{\partial \varepsilon}  $$
    
 

now the value $\frac{\partial x_1}{\partial \varepsilon} = \eta(t, \varepsilon) $ from above 
and the value $\frac{\partial \dot x_1 }{\partial \varepsilon} = \frac{\partial \eta(t, \varepsilon) }{\partial t} $ is defined from above, and represents the derivative of $\eta(t)$ with respect to $t$. 

<span style = "color:red"> <div style="float:left"><img src = "https://drive.google.com/uc?export=view&id=0B-Q-Pf4fc8LdUnpSdUo3emczWjA" style="width:15%;height:15%"></img></div> Hang on, you might ask ... how is the derivative of $\dot x_1$ with respect to $\varepsilon$ equal to the derivative of $\eta(t,\varepsilon)$ with respect to time? Well the answer is that <br>
$$
\frac{d\dot x}{d\varepsilon} = \frac{d \frac{d x}{dt}} {d\varepsilon} = \frac{d (d x)}{d t d\varepsilon} = \frac{d}{dt}\frac{d x}{d \varepsilon} = \frac{d}{dt}\eta(t,\varepsilon) 
$$

</span>        
    
from inspection of the total differential, the values above can be substituted and the total differential becomes <br>
   
$$ \frac{\partial f}{\partial \varepsilon} = \frac{\partial f}{ \partial x_1} \eta(t,\varepsilon) + \frac{ \partial f}{ \partial \dot x_1}\frac{\partial\eta(t,\varepsilon)}{\partial t} $$    
  
this total differential is substituted into the integral to represent the integral of the infinitesimal variation of $J$. <br>

$$ \delta J(f) = \int_a^b \left ( \displaystyle  \frac{\partial f}{ \partial x_1} \eta(t,\varepsilon) + \frac{ \partial f}{ \partial \dot x_1}\frac{\partial\eta(t,\varepsilon)}{\partial t}\right) dt  $$ 
   
and this is done in pieces using the fact that the integral of a sum is the sum of integrals... 
    
$$ \displaystyle \delta J(f) = \int_a^b \left ( \displaystyle  \frac{\partial f}{ \partial x_1} \eta(t,\varepsilon)\right) dt + \int_a^b \left(\frac{ \partial f}{ \partial \dot x_1}\frac{\partial\eta(t,\varepsilon)}{\partial t}\right) dt  $$ 
   
    
The second term is integrated by parts, using the relation $ u dv = d(uv) - v du $ with $v = \eta(t)$ and integrating  to remove the derivative of $\eta(t,\varepsilon)$ then...
   
$$ \delta J(f) = \int_a^b \left ( \displaystyle  \frac{\partial f}{ \partial x_1} \eta(t,\varepsilon)\right) dt + \left[\eta(t,\varepsilon) \frac{ \partial f}{ \partial \dot x_1}\right]_b^a - \int_a^b \left( \left(\frac{\partial }{\partial t}\frac{ \partial f}{ \partial \dot x_1}\right)\eta(t,\varepsilon)\right) dt  $$ <br> 
    
the value of $\eta(t)$ was defined to be zero at $t=a$ and $t=b$ above, since the curves share the same start and end points, so the term $\left[\eta(t,\varepsilon) \frac{ \partial f}{ \partial \dot x_1}\right]_b^a$ vanishes and the integral can be recombined to factorize $\eta (t,\varepsilon)$ then

$$ \delta J(f) = \int_a^b \left ( \displaystyle  \frac{\partial f}{ \partial x_1} -\left(\frac{\partial }{\partial t}\frac{ \partial f}{ \partial \dot x_1}\right)\right) \eta(t,\varepsilon) dt   $$ 
    
Now this is where the <b> fundamental Lemma of the Calculus of Variations </b> is applied. The lemma states that if the equality is satisfied 

$$A= \int_a^b H(x)G(x)dx = 0  $$

For all smooth continuous functions H defined on (a,b), then $G(x)$ is identically zero. <br>

Applying this Lemma to the integral shows that \(\eta(t)\) the error involved is identically zero, and we wish to solve the term inside the brackets <br>

$$
\frac{\partial f}{ \partial x_1} -\frac{\partial }{\partial t}\left(\frac{ \partial f}{ \partial \dot x_1}\right) = 0
$$

this term is known as the Lagrangian, and is one of the most important equations in Classical Mechanics, the basis for Lagrangian Dynamics, and the primary method for solving problems in the Calculus of Variations. This problem was first solved by Joseph-Louis Lagrange in the 1750s. This result can be extended to higher dimensions. The Lagrangian can also be expressed as 
    
$$ L = T-V $$ 

Where T is the kinetic energy (usually $\frac{1}{2} m v^2$ where v is velocity and m is mass. The value V is the potential energy, and is usually $mgh$ or mass times gravity (acceleration) times height. 
    
  


	
### Calculus Of Variations
    
This involves the study of finding an extremum (minima or maxima) of functionals. From the section on the Euler-Lagrange equation we find that to minimize the functional, <br>
    
$$
\displaystyle J(x)= \int_a^b L(x,\ \dot x) dt 
$$
   
we can use the Euler-Lagrange equation, or the Lagrangian <br>
   
$$
   \displaystyle \frac{\partial L}{\partial x}-\frac{\partial}{\partial t} \left(\frac{\partial L}{\partial \dot x}\right) = 0 \qquad (1)
$$

or 
$$
   \displaystyle \frac{\partial L}{\partial x}= \frac{\partial}{\partial t}\left(\frac{\partial L}{\partial \dot x}\right ) \qquad (2)
$$
    
The functional in this relation is not explicitly a function of \(t\) and so for these special cases the Beltrami identity also applies ... <br>
$$
\displaystyle \dot x \frac{\partial L}{\partial \dot x} - L = C  \qquad (3)
$$
where $C$ is a constant. To prove this we start with

$$
\displaystyle \frac{\partial }{\partial t}\left(\frac{\partial L}{\partial \dot x}\right) =    \frac{\partial }{\partial \dot x}\left(\frac{\partial L}{\partial x} \frac{\partial x}{\partial t} \right) 
    \qquad (4)
$$

Then by the product rule ... 
    
$$
\frac{\partial }{\partial \dot x}\left(\frac{\partial L}{\partial x} \frac{\partial x}{\partial t} \right) = \frac{\partial^2 L}{\partial x \partial \dot x}\frac{\partial x}{\partial t} + \frac{\partial L}{\partial x} \frac{\partial \dot x}{ \partial \dot x} = \frac{\partial^2 L}{\partial x \partial \dot x}\frac{\partial x}{\partial t} + \frac{\partial L}{\partial x} \qquad (5) 
$$
    

we can then derive the equation by recognizing that <br>
    
$$
\displaystyle \frac{\partial }{\partial x}\left (\dot x \frac{\partial L}{\partial \dot x}\right) = \frac{\partial \dot x}{\partial x} \frac{\partial L}{\partial \dot x} + \dot x \frac{\partial^2 L}{\partial \dot x \partial x}
$$    

After cancelling the $\partial x$'s from the first term this is the sane as the RHS of equation (5) so putting that into (1) and integrating with respect to X gives the Beltrami identity.
$$
\displaystyle \int\frac{\partial L}{\partial x}dx - \int\left(\frac{\partial^2 L}{\partial x \partial \dot x}\frac{\partial x}{\partial t} + \frac{\partial L}{\partial x}\right)dx = 0
$$

and effectively swapping the subtraction sign by subtracting both terms from both sides (moving them to the other side of the equality) and using the product rule :<br>

$$
\displaystyle \dot x \frac{\partial L}{\partial \dot x} - L = const
$$

where $const$ is the constant of integration. This suggests a conservation law, however this isn't proven.

### Shortest Path Problem
    
To resolve the simple problem of the shortest path between two points $(x_1,y_1)$ and $(x_2,y_2)$ in a plane with a function $y=f(x)$ we can state simply that the differential element $\Delta s$ of a line has the length $\Delta s = \sqrt{\Delta x^2+\Delta y^2}$, now usually the products of differentials (infinitesimals) are removed or ignored because they become too small, however in this case the follwing holds

$$\displaystyle \Delta s= \sqrt{ \Delta x^2 + \Delta y^2  } = \sqrt{ \left( 1+ \left(\frac{\Delta y}{\Delta x}\right)^2\right)\Delta x^2 } =\sqrt{  1+ \left(\frac{\Delta y}{\Delta x}\right)^2}\Delta x $$

Hence since they are in the division form we see that in the limit they become $dy/dx$ and 
    
$$\displaystyle \lim_{\Delta x,\Delta y\to 0}\sum_{n=0}^{N=(x_2-x_1)/\Delta x} ds_n= \lim_{\Delta x,\Delta y\to 0}\sum_{n=0}^{N \to \infty} \sqrt{ 1 + \left(\frac{\Delta y_n}{\Delta x_n}\right)^2  }\Delta x_n = \int_{x_1}^{x_2}\sqrt{ 1 + \left(\frac{d y}{d x}\right)^2  }d x$$
   
This term wihin the integrand that we have already solved for the infinitesimal arc length can be solved using the Euler-Lagrange equation. Since the integral does not depend on $x$ the second term is the only term that needs to be computed. Also the variable for the horizontal axis is \(x\) and not time \(t\) in this case, so the term becomes
   
$$ \frac{\partial}{\partial x} \left(\frac{\partial s}{\partial y'}\right) = 0$$
    

First computing the derivative with respect to $dy/dx$ gives :
    
$$ \frac{\partial s}{\partial y'} = \frac{d}{dy'} \left(\sqrt{ 1 + \left(\frac{d y}{d x}\right)^2  } \right)$$
   
making the substitution 
   
$$u = 1 + \left(\frac{d y}{d x}\right)^2 $$ 
   
then 
   
$$\frac{du}{dy'} = 2y'$$ 
   
so 
   
$$\frac{ds}{dy'} = \frac{ds}{du}\frac{du}{dy'}$$ 

so ...
    
$$ \displaystyle \frac{\partial s}{\partial y'} = \frac{ds}{du}\frac{du}{dy'}=\left(\frac{1}{2}u^{-\frac{1}{2}}\right)\left(2y'\right) = \frac{y'}{\sqrt{ 1 + \left(\frac{d y}{d x}\right)^2  }}$$
   

and the derivative of this with respect to $x$ is given by 
   
$$\frac{d}{dx}\frac{\partial s}{\partial y'} = \frac{y''}{\left( 1 + \left(\frac{d y}{d x}\right)^2  \right) ^{3/2}} $$ 
       
since the demoninator is always positive the differential equation to solve is the numerator, and this means 


$$ y'' =0$$
   
so the twice differentiable function that equals zero can only be the class of straight lines 

$$y= mx + b$$
   
### Best Half Pipe

What is the optimal shape for a half-pipe (surface from skateboarding or snowboarding)? The question can best be formulated as "what shape will allow an object to travel between two points A and B separated a vertical distance h and a horizontal distance l in the least amount of time?"
    <img src="public/images/ramp_path.png"></img> <br>
  
The exact dimensions of $h$ and $l$ are not specified, but we can assume that we need to find a curve that maximizes velocity, and so therefore may use gravity to accelerate. The method of solution is as follows <br>

Setting up the conservation of energy equation $E=K+V=const$ with kinetic energy $K=\frac{1}{2}mv^2$ and potential energy $V=mgh$ then we can solve for $v$ the velocity by observing that the [const] value is equal to the potential energy at the start, and so for maximum kinetic energy $v=\sqrt{2gh}$. Then the optimal curve for the fall can be calculated as the sum of infinitesimal arc lengths, from above we have the result that $d s = \sqrt{1+(y')^2} d x$ and the time of travel is $\frac{ds}{v} = dt$, so the integral can be expressed as 

$$ I = \int_A^B dt = \int_A^B \frac{ds}{v} = \frac{1}{\sqrt{2g}}\int_A^B  \frac{\sqrt{1+(y')^2}}{\sqrt y} d x $$
   
This problem is known as the *brachistochrone problem*. Since $x$ does not explicitly appear in the function, the Beltrami identity is applied. 
  
$$ y' \frac{\partial f}{\partial y'} - f = y' \frac{d}{dy'} \left( \frac{\sqrt{1+(y')^2}}{\sqrt y} \right) - \frac{\sqrt{1+(y')^2}}{\sqrt y} = const $$

and so for the term involving $d/dy'$ the substitution is made $u=1+(y')^2$ and $\frac{df}{du} = \frac{1}{2}u^{-1/2}$ and  $\frac{du}{dy'} = 2y'$
   
$$ y' \left(   \frac{1}{2} ( 1+(y')^2 )^{-1/2} \right)(2y') ( \frac{1}{\sqrt y} ) - \frac{\sqrt{1+(y')^2}}{\sqrt y} = const $$
   
and so

$$ \frac{(y')^2}{\sqrt{ 1+(y')^2} \sqrt{y}}  - \frac{\sqrt{1+(y')^2}}{\sqrt y} = \frac{(y')^2 - 1-(y')^2 }{\sqrt{ 1+(y')^2} \sqrt{y}} =  \frac{ - 1 }{\sqrt{ 1+(y')^2} \sqrt{y}} = const $$
   
so solving for $y'$ we obtain 

$$ \frac{ - 1 }{ C \sqrt{y}} = \sqrt{ 1+(y')^2} $$

then
$$ \frac{ 1 }{ C^2 y} =  1+(y')^2 $$
and
$$ \frac{ 1 }{ C^2 y} -1=  (y')^2 \\
$$
etc

$$ \frac{ 1 -C^2 y }{ C^2 y} =  (y')^2 \\
$$
then
$$
\sqrt{ \frac{ \frac{1}{C^2} - y }{  y}} =  y' \\
$$<br>

and so to seperate variables and integrate, the substitution for $\tan (\theta)$ can be used 

$$ \tan(\theta) = \sqrt{\frac{y}{k-y}} $$

and so

$$
(k-y) \tan^2(\theta) = y $$
and
$$ k\tan^2(\theta)  = y(1+\tan^2(\theta))\\
$$

and therefore 
$$y = \frac{k \tan^2(\theta)}{(1+\tan^2(\theta))} $$

using the fact that $1+\tan^2(\theta) = \sec^2(\theta) = \frac{1}{\cos^2(\theta)}$ the result was
   
$$ y = \frac{k \tan^2(\theta)}{\frac{1}{\cos^2(\theta)}} = k\frac{\frac{\sin^2(\theta)}{\cos^2(\theta)}}{\frac{1}{\cos^2(\theta)}} = k \sin^2(\theta) $$

resolving $d\theta/dx = (d\theta/dy)(dy/dx)=1/(2k\sin^2(\theta)  $ and setting $dx = 2k\sin^2(\theta) d\theta $ and integrating gives 

$$
x = 2k\left( \frac{\theta}{2}-\frac{\sin(2\theta)}{4}\right)
$$

and so substituting $a = k/2$ and $\phi = 2\theta$ with the assumption that at point A the coordinates x and y are zero therefore the angle $\theta$ is zero, and so 

$$
x=a(\phi - \sin(\phi))$$
and
$$y=a(1-\cos(\phi))
$$

and these are the equations of a cycloid.

### Newtons Law In Lagrangian Vector form

Given an intial velocity for a thrown stone we solve for the space curve <br>

$$
\mathbf v_1=10\  \mathbf i + 5\  \mathbf j +0\mathbf k
$$

we have the kinetic energy  (mass =1  throughout for simplicity )

$$
K=\frac{\mathbf v_1\cdot\mathbf v_1}{2} $$

and the potential energy from gravity acting on the y axis (in this case) 

$$
U=g\mathbf r_1=(0,g,0)
$$

forming the Lagrangian 

$$
L=K-U = \frac{\mathbf v_1\cdot \mathbf v_1}{2}-g\mathbf r $$

we then compute the derivatives <br>

$$
\frac{d}{dt}\frac{\partial L}{\partial \mathbf v}=\frac{d}{dt}\mathbf v_1 =  \mathbf{\ddot r}
$$

since $\frac{d (\mathbf{v\cdot v})}{d\mathbf v} = \mathbf v\cdot\frac{d\mathbf v}{d\mathbf v} + \frac{d\mathbf v}{d\mathbf v}\cdot\mathbf v = 2\mathbf v$ 
we have
$$
\frac{\partial L}{\partial \mathbf r} = -g\mathbf j$$
and so the acceleration is

$$\mathbf{\ddot r} = -g \mathbf j $$

assuming the stone is released from the arm at a height of 2m 

$$
r_0=(0,2,0)
$$

integrating twice gives first velocity so we can substitute intial velocity $\mathbf v_1$ and finally displacement so we can substitute the intial height vector 

$$
r = \mathbf r_0 + \mathbf v_1 t - \frac{1}{2}g\ t^2\ \mathbf j
$$
   
solve for the height $y = 2+ 5t- 4.9t^2=2+(5-4.9t)t$ so that $5/4.9 = 1.02s$ before the stone reaches 2m above the ground again. The final 2m is calculated using the quadratic formula 

$$
\frac{-b\pm\sqrt{b^2-4ac}}{2a}=\frac{-5\pm\sqrt{25+4\times4.9\times 2}}{9.8} \frac{-5\mp\sqrt{64.2}}{9.8}=\frac{-5+8.01}{9.8}=0.307 $$

using only the positive root, (the negative root gives the same answer as adding these other two results). Now adding the two gives the time of flight 

$$
time=1.02 + 0.307=1.327\\
$$
this is easily checked using our y axis equation 

$$
y = 2+ (5-4.9\times 1.327)\times 1.327 = 2+(5-6.5023)\times 1.327 = 2+(-1.5023)\times 1.327 = 2-1.99 
$$

close enough to zero, that means hitting the ground. From our velocity equation we can see that the $x$ value is 10, so 

$$
x=10t=10\times 1.327 =13.27 meters.
